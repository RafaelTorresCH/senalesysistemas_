## Comparacin entre las formas de la serie de Fourier

La serie de Fourier es una herramienta fundamental en el anlisis de seales, ya que permite representar cualquier seal peridica como una suma de funciones sinusoidales. Existen tres formas principales para expresar esta serie: **forma trigonomtrica**, **forma exponencial** y **forma compacta**. Aunque todas describen lo mismo, difieren en la forma matemtica de escribir la seal y en la interpretacin de los coeficientes.

### 1. Serie de Fourier Trigonomtrica

Esta forma representa la seal como una suma infinita de senos y cosenos. La frmula general es:

$ x(t) = a_0 + \sum_{n=1}^{\infty} \left( a_n \cos(n \omega_0 t) + b_n \sin(n \omega_0 t) \right) $

- $a_0$ es el valor promedio (componente DC) de la seal.
- $a_n$ y $b_n$ son coeficientes reales que indican la amplitud de cada componente armnica.
- $\omega_0 = \frac{2\pi}{T}$ es la frecuencia fundamental.

Esta forma es fcil de interpretar porque cada trmino corresponde a una frecuencia especfica, con una amplitud y una fase definida.

### 2. Serie de Fourier Exponencial

En esta versin, la seal se expresa usando exponenciales complejas, lo cual facilita muchos anlisis matemticos:

$ x(t) = \sum_{n=-\infty}^{\infty} C_n e^{j n \omega_0 t} $

- Los $C_n$ son coeficientes complejos que contienen toda la informacin de amplitud y fase.
- Esta forma es ms compacta y general que la trigonomtrica, y se usa mucho en teora de sistemas y procesamiento de seales.

### 3. Forma Compacta

El trmino "forma compacta" suele referirse simplemente a la forma exponencial, ya que es una versin ms condensada y algebraicamente manejable. Se basa en el uso de nmeros complejos para representar la seal de forma ms eficiente.

---

### Semejanzas

- Todas representan seales peridicas.
- Permiten descomponer una seal en sumas de funciones ms simples (senos, cosenos o exponenciales).
- Los coeficientes de cada forma estn relacionados y se pueden convertir de una forma a otra.

### Diferencias

| Caracterstica          | Trigonomtrica                            | Exponencial                                | Compacta                             |
|-------------------------|--------------------------------------------|---------------------------------------------|---------------------------------------|
| Funciones base          | Senos y cosenos                           | Exponenciales complejas                     | Equivalente a la exponencial          |
| Coeficientes            | Reales ($a_n$, $b_n$)                     | Complejos ($C_n$)                           | Complejos                             |
| Interpretacin fsica   | Ms intuitiva                             | Ms tcnica, til para anlisis matemtico  | Similar a la exponencial              |
| Simetra                | Explcita                                 | Implcita ($C_{-n} = C_n^*$ para seales reales) | Idem exponencial                  |

## Transformada de Fourier, DTFT y DFT: semejanzas y diferencias

Despus de entender las series de Fourier para seales peridicas, es importante conocer cmo se analiza una seal **no peridica** o **discreta**. Aqu es donde aparecen la **Transformada de Fourier (FT)**, la **Transformada de Fourier en tiempo discreto (DTFT)** y la **Transformada Discreta de Fourier (DFT)**. Aunque tienen nombres parecidos, cada una se usa en un contexto diferente.

### 1. Transformada de Fourier (FT)

Se aplica a **seales continuas en el tiempo** que pueden ser **no peridicas**. Convierte la seal del dominio del tiempo al dominio de la frecuencia continua.

$ X(f) = \int_{-\infty}^{\infty} x(t) e^{-j 2\pi f t} dt $

- $x(t)$ es la seal en el tiempo.
- $X(f)$ es la representacin en frecuencia.
- El espectro resultante es **continuo**.

Se usa mucho en fsica, comunicaciones y procesamiento de seales analgicas.

### 2. Transformada de Fourier en Tiempo Discreto (DTFT)

Se aplica a seales **discretas en el tiempo pero infinitas**. Tambin entrega un espectro **continuo**, pero peridico.

$ X(e^{j\omega}) = \sum_{n=-\infty}^{\infty} x[n] e^{-j \omega n} $

- $x[n]$ es la seal discreta.
- $X(e^{j\omega})$ es su representacin en frecuencia, peridica en $\omega$ con periodo $2\pi$.

La DTFT es muy til en teora de procesamiento digital de seales.

### 3. Transformada Discreta de Fourier (DFT)

Se aplica a seales **discretas y de duracin finita**. A diferencia de la DTFT, el resultado es tambin **discreto** (en frecuencia).

$ X[k] = \sum_{n=0}^{N-1} x[n] e^{-j \frac{2\pi}{N} kn} $

- $x[n]$ es la seal de entrada, con $N$ muestras.
- $X[k]$ representa las componentes de frecuencia discretas.
- Se implementa fcilmente en computador y es base para algoritmos como la FFT.

### Comparacin general

| Aspecto                         | FT                           | DTFT                                | DFT                                |
|----------------------------------|-------------------------------|--------------------------------------|-------------------------------------|
| Tipo de seal                    | Continua, infinita           | Discreta, infinita                  | Discreta, finita                    |
| Resultado (espectro)            | Continuo                     | Continuo, pero peridico             | Discreto y peridico                |
| Aplicacin                      | Seales analgicas           | Seales digitales tericas          | Seales digitales reales (cmputo) |
| Implementacin en computador    | No                           | No directamente                     | S                                  |

## Qu es la FFT?

Aqu es donde entra el algoritmo **Fast Fourier Transform (FFT)**, que explicaremos en el siguiente punto.


| Concepto                            | Descripcin                                                                                                                                     |
|-------------------------------------|-------------------------------------------------------------------------------------------------------------------------------------------------|
| **DFT (Transformada Discreta de Fourier)** | Calcula las componentes en frecuencia de una seal digital. Requiere muchas operaciones.                                                        |
| **Costo computacional DFT**         | $O(N^2)$, porque necesita $N$ sumas para cada uno de los $N$ resultados.                                                                        |
| **FFT (Transformada Rpida de Fourier)** | Es un algoritmo eficiente para calcular la DFT. Usa un enfoque de dividir y conquistar (por ejemplo, el algoritmo Cooley-Tukey).              |
| **Costo computacional FFT**         | $O(N \log_2 N)$, mucho ms eficiente que la DFT directa.                                                                                        |
| **Ventaja principal de la FFT**     | Reduce significativamente el tiempo de clculo, lo que la hace til en procesamiento de seales, audio, imagen, etc.                            |
| **Mtodo principal de la FFT**      | Divide la seal en partes (pares e impares), calcula la DFT de cada parte y luego combina los resultados.                                       |

## Transformadas de Fourier  Paso a Paso

### a)   $ x(t) = e^{-a|t|}, \; a > 0 $

$$
X(f) = \int_{-\infty}^{\infty} e^{-a|t|} e^{-j 2\pi f t} dt
$$

Partimos en dos integrales:

$$
X(f) = \int_{-\infty}^{0} e^{(a - j2\pi f)t} dt + \int_{0}^{\infty} e^{-(a + j2\pi f)t} dt
$$

Evaluamos ambas:

$$
X(f) = \frac{1}{a - j2\pi f} + \frac{1}{a + j2\pi f}
= \frac{2a}{a^2 + (2\pi f)^2}
$$

---

### b) $ x(t) = \cos(w_c t) $

Usamos la forma exponencial:

$$
\cos(w_c t) = \frac{1}{2}(e^{j w_c t} + e^{-j w_c t})
$$

Entonces:

$$
X(f) = \frac{1}{2} \left[ \delta\left(f - \frac{w_c}{2\pi}\right) + \delta\left(f + \frac{w_c}{2\pi}\right) \right]
$$

---

### c) $ x(t) = \sin(w_s t) $

Forma exponencial:

$$
\sin(w_s t) = \frac{1}{2j}(e^{j w_s t} - e^{-j w_s t})
$$

Transformada:

$$
X(f) = \frac{1}{2j} \left[ \delta\left(f - \frac{w_s}{2\pi}\right) - \delta\left(f + \frac{w_s}{2\pi}\right) \right]
$$

---

### d) $x(t) = f(t)\cos(w_c t) $

No se puede resolver sin conocer la forma exacta de \( f(t) \).  
En general, este caso implica una convolucin en frecuencia, pero como **no se deben usar propiedades**, no se puede continuar directamente.

---

### e) $ x(t) = e^{-a|t|^2}, \; a > 0 $

Transformada de una gaussiana:

$$
X(f) = \int_{-\infty}^{\infty} e^{-a t^2} e^{-j 2\pi f t} dt
$$

Resultado conocido:

$$
X(f) = \sqrt{\frac{\pi}{a}} \, e^{-\frac{(\pi f)^2}{a}}
$$

---

### f) $ x(t) = A \, \text{rect}_d(t) $

La funcin \( \text{rect}_d(t) \) est definida en el intervalo \( [-d/2, d/2] \):

$$
X(f) = \int_{-d/2}^{d/2} A \, e^{-j 2\pi f t} dt
$$

Evaluamos:

$$
X(f) = A \cdot \frac{\sin(\pi f d)}{\pi f} = A d \cdot \text{sinc}(f d)
$$

Donde:

$$
\text{sinc}(x) = \frac{\sin(\pi x)}{\pi x}
$$

#Demostracin Analtica (Sistema 1):
Linealidad: La presencia del trmino recursivo $-y[n-1]$ y su interaccin con las entradas hacen que el sistema no sea lineal en su forma general, especialmente si las condiciones iniciales no son triviales o se asume el sistema en reposo de manera estricta. La superposicin no se mantiene universalmente.
Invariancia en el Tiempo: El trmino $-y[n-1]$ tambin viola la invariancia temporal. Un retraso en la entrada afectara la secuencia de $y[n-1]$ de forma que la salida retrasada no sera simplemente la salida original retrasada

Sistema 1: No es un SLIT.

```python

## 1. Sistema $y[n] = x[n]/3 + 2x[n-1] - y[n-1]$

### **Simulacin en Python:**

import numpy as np
import matplotlib.pyplot as plt

def simulate_system1(x, initial_y=0):
    y = np.zeros_like(x, dtype=float)
    y_prev = initial_y
    for n in range(len(x)):
        x_n_minus_1 = x[n-1] if n > 0 else 0
        y[n] = x[n]/3 + 2 * x_n_minus_1 - y_prev
        y_prev = y[n]
    return y

N = 20 # Nmero de muestras

# --- Linealidad: Aditividad ---
x1 = np.zeros(N); x1[5:10] = 1
x2 = np.zeros(N); x2[12:15] = 0.5
y1 = simulate_system1(x1)
y2 = simulate_system1(x2)
y_sum_input = simulate_system1(x1 + x2)
y_sum_outputs = y1 + y2
print(f"Sistema 1 - Es aditivo? {np.allclose(y_sum_input, y_sum_outputs)}") # Esperado: False

# --- Linealidad: Homogeneidad ---
a = 3
x_hom = np.zeros(N); x_hom[5:10] = 1
y_x_hom = simulate_system1(x_hom)
y_ax_hom = simulate_system1(a * x_hom)
print(f"Sistema 1 - Es homogneo? {np.allclose(a * y_x_hom, y_ax_hom)}") # Esperado: False

# --- Invariancia en el Tiempo ---
k0 = 5
x_orig_ti = np.zeros(N); x_orig_ti[5:10] = 1
x_shifted_ti = np.zeros(N)
if N - k0 > 0: x_shifted_ti[5+k0 : 10+k0] = 1
y_orig_ti = simulate_system1(x_orig_ti)
y_shifted_input_ti = simulate_system1(x_shifted_ti)
y_orig_shifted_ti = np.zeros(N)
if N - k0 > 0: y_orig_shifted_ti[k0:] = y_orig_ti[:-k0]
print(f"Sistema 1 - Es invariante en el tiempo? {np.allclose(y_shifted_input_ti, y_orig_shifted_ti)}") # Esperado: False

plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.stem(range(N), y_sum_input, linefmt='b-', markerfmt='bo', basefmt=' ')
plt.stem(range(N), y_sum_outputs, linefmt='r--', markerfmt='rx', basefmt=' ')
plt.title('Aditividad: y[x1+x2] (Azul) vs y1+y2 (Rojo)')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.stem(range(N), y_ax_hom, linefmt='b-', markerfmt='bo', basefmt=' ')
plt.stem(range(N), a * y_x_hom, linefmt='r--', markerfmt='rx', basefmt=' ')
plt.title('Homogeneidad: y[a*x] (Azul) vs a*y[x] (Rojo)')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.stem(range(N), y_shifted_input_ti, linefmt='b-', markerfmt='bo', basefmt=' ')
plt.stem(range(N), y_orig_shifted_ti, linefmt='r--', markerfmt='rx', basefmt=' ')
plt.title('Invariancia: y[x[n-k0]] (Azul) vs y[x][n-k0] (Rojo)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

#2. Sistema $y[n] = \sum_{k=-\infty}^{n} x^2[k]$
Este sistema discreto involucra una operacin no lineal ($x^2[k]$).

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_system2(x):
    y = np.zeros_like(x, dtype=float)
    cumulative_sum_sq = 0
    for n in range(len(x)):
        cumulative_sum_sq += x[n]**2
        y[n] = cumulative_sum_sq
    return y

N = 20

# --- Linealidad: Aditividad ---
x1 = np.zeros(N); x1[5] = 2
x2 = np.zeros(N); x2[6] = 3
y1 = simulate_system2(x1)
y2 = simulate_system2(x2)
y_sum_input = simulate_system2(x1 + x2)
y_sum_outputs = y1 + y2
print(f"Sistema 2 - Es aditivo? {np.allclose(y_sum_input, y_sum_outputs)}") # Esperado: False

# --- Linealidad: Homogeneidad ---
a = 2
x_hom2 = np.zeros(N); x_hom2[5] = 2
y_x_hom2 = simulate_system2(x_hom2)
y_ax_hom2 = simulate_system2(a * x_hom2)
print(f"Sistema 2 - Es homogneo? {np.allclose(a * y_x_hom2, y_ax_hom2)}") # Esperado: False

# --- Invariancia en el Tiempo ---
k0 = 3
x_orig_ti2 = np.zeros(N); x_orig_ti2[5:8] = 1
x_shifted_ti2 = np.zeros(N)
if N - k0 > 0: x_shifted_ti2[5+k0 : 8+k0] = 1
y_orig_ti2 = simulate_system2(x_orig_ti2)
y_shifted_input_ti2 = simulate_system2(x_shifted_ti2)
y_orig_shifted_ti2 = np.zeros(N)
if N - k0 > 0: y_orig_shifted_ti2[k0:] = y_orig_ti2[:-k0]
print(f"Sistema 2 - Es invariante en el tiempo? {np.allclose(y_shifted_input_ti2, y_orig_shifted_ti2)}") # Esperado: True

plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.stem(range(N), y_sum_input, linefmt='b-', markerfmt='bo', basefmt=' ')
plt.stem(range(N), y_sum_outputs, linefmt='r--', markerfmt='rx', basefmt=' ')
plt.title('Aditividad: y[x1+x2] (Azul) vs y1+y2 (Rojo)')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.stem(range(N), y_ax_hom2, linefmt='b-', markerfmt='bo', basefmt=' ')
plt.stem(range(N), a * y_x_hom2, linefmt='r--', markerfmt='rx', basefmt=' ')
plt.title('Homogeneidad: y[a*x] (Azul) vs a*y[x] (Rojo)')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.stem(range(N), y_shifted_input_ti2, linefmt='b-', markerfmt='bo', basefmt=' ')
plt.stem(range(N), y_orig_shifted_ti2, linefmt='r--', markerfmt='rx', basefmt=' ')
plt.title('Invariancia: y[x[n-k0]] (Azul) vs y[x][n-k0] (Rojo)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Demostracin Analtica (Sistema 2):
Linealidad: La operacin $x^2[k]$ rompe la linealidad.
Aditividad: $\sum (x_1+x_2)^2 \neq \sum x_1^2 + \sum x_2^2$ debido al trmino cruzado $2x_1x_2$.
Homogeneidad: $\sum (ax)^2 = a^2 \sum x^2 \neq a \sum x^2$ (a menos que $a=0$ o $a=1$). Por lo tanto, no es lineal.
Invariancia en el Tiempo: El operador de suma y la potencia no dependen del instante de tiempo absoluto $n$, solo del ndice relativo $k$. Un desplazamiento en la entrada $x[n-k_0]$ resulta en un desplazamiento idntico en la salida. Por lo tanto, s es invariante en el tiempo.
Conclusin para el Sistema 2: No es un SLIT (por no ser lineal).

3. Sistema $y[n] = \text{median}(x[n-1], x[n], x[n+1])$
Este es un sistema discreto de filtrado de mediana con ventana de 3. La operacin de mediana es no lineal.

Simulacin en Python:

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_system3(x):
    y = np.zeros_like(x, dtype=float)
    x_padded = np.pad(x, (1, 1), 'constant', constant_values=0)
    for n in range(len(x)):
        window = [x_padded[n], x_padded[n+1], x_padded[n+2]]
        y[n] = np.median(window)
    return y

N = 10

# --- Linealidad: Aditividad ---
x1 = np.zeros(N); x1[3] = 10
x2 = np.zeros(N); x2[4] = 1
y1 = simulate_system3(x1)
y2 = simulate_system3(x2)
y_sum_input = simulate_system3(x1 + x2)
y_sum_outputs = y1 + y2
print(f"Sistema 3 - Es aditivo? {np.allclose(y_sum_input, y_sum_outputs)}") # Esperado: False

# --- Linealidad: Homogeneidad ---
a = 2
x_hom3 = np.zeros(N); x_hom3[3:6] = [1, 5, 2]
y_x_hom3 = simulate_system3(x_hom3)
y_ax_hom3 = simulate_system3(a * x_hom3)
print(f"Sistema 3 - Es homogneo? {np.allclose(a * y_x_hom3, y_ax_hom3)}") # Esperado: True (para a>=0) o False si a<0 y elementos cruzan 0.

# --- Invariancia en el Tiempo ---
k0 = 2
x_orig_ti3 = np.zeros(N); x_orig_ti3[3:6] = [1, 5, 2]
x_shifted_ti3 = np.zeros(N)
if N - k0 > 0: x_shifted_ti3[3+k0 : 6+k0] = [1, 5, 2]
y_orig_ti3 = simulate_system3(x_orig_ti3)
y_shifted_input_ti3 = simulate_system3(x_shifted_ti3)
y_orig_shifted_ti3 = np.zeros(N)
if N - k0 > 0: y_orig_shifted_ti3[k0:] = y_orig_ti3[:-k0]
print(f"Sistema 3 - Es invariante en el tiempo? {np.allclose(y_shifted_input_ti3, y_orig_shifted_ti3)}") # Esperado: True

plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.stem(range(N), y_sum_input, linefmt='b-', markerfmt='bo', basefmt=' ')
plt.stem(range(N), y_sum_outputs, linefmt='r--', markerfmt='rx', basefmt=' ')
plt.title('Aditividad: y[x1+x2] (Azul) vs y1+y2 (Rojo)')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.stem(range(N), y_ax_hom3, linefmt='b-', markerfmt='bo', basefmt=' ')
plt.stem(range(N), a * y_x_hom3, linefmt='r--', markerfmt='rx', basefmt=' ')
plt.title('Homogeneidad: y[a*x] (Azul) vs a*y[x] (Rojo)')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.stem(range(N), y_shifted_input_ti3, linefmt='b-', markerfmt='bo', basefmt=' ')
plt.stem(range(N), y_orig_shifted_ti3, linefmt='r--', markerfmt='rx', basefmt=' ')
plt.title('Invariancia: y[x[n-k0]] (Azul) vs y[x][n-k0] (Rojo)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Demostracin Analtica (Sistema 3):
Linealidad:
Aditividad: La mediana de una suma no es la suma de las medianas (ej., median(0, 10, 5) = 5, pero median(0,10,0)+median(0,0,5) = 0+0=0). No es aditivo.
Homogeneidad: $\text{median}(a \cdot \text{ventana}) = a \cdot \text{median}(\text{ventana})$. S es homogneo. Dado que falla la aditividad, no es lineal.
Invariancia en el Tiempo: La operacin de mediana sobre una ventana es intrnsecamente independiente del tiempo absoluto. Un desplazamiento de la entrada resulta en el mismo desplazamiento de la salida. S es invariante en el tiempo.
Conclusin para el Sistema 3: No es un SLIT (por no ser lineal).


# 4. Sistema $y(t) = Ax(t) + B$; $A, B \in \mathbb{R}$

Este es un sistema continuo con un trmino constante aditivo.

Simulacin en Python:

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_system4(x, A, B):
    return A * x + B

A_val = 2
B_val = 5 # Valor no nulo para la constante

# --- Linealidad: Aditividad ---
t = np.linspace(0, 10, 100)
x1_cont = np.sin(t)
x2_cont = np.cos(t)
y_sum_input_cont = simulate_system4(x1_cont + x2_cont, A_val, B_val)
y_sum_outputs_cont = simulate_system4(x1_cont, A_val, B_val) + simulate_system4(x2_cont, A_val, B_val)
print(f"Sistema 4 - Es aditivo? {np.allclose(y_sum_input_cont, y_sum_outputs_cont)}") # Esperado: False si B != 0

# --- Linealidad: Homogeneidad ---
a_scale = 3
x_hom4 = np.sin(t)
y_ax_hom4 = simulate_system4(a_scale * x_hom4, A_val, B_val)
y_a_times_y_x_hom4 = a_scale * simulate_system4(x_hom4, A_val, B_val)
print(f"Sistema 4 - Es homogneo? {np.allclose(y_ax_hom4, y_a_times_y_x_hom4)}") # Esperado: False si B != 0 y a != 1

# --- Invariancia en el Tiempo ---
k0_time = 2 # Retraso en segundos
dt = t[1] - t[0]
k0_samples = int(k0_time / dt)

x_orig_ti4 = np.sin(t)
y_orig_ti4 = simulate_system4(x_orig_ti4, A_val, B_val)

x_shifted_ti4 = np.zeros_like(t)
x_shifted_ti4[k0_samples:] = x_orig_ti4[:-k0_samples]

y_shifted_input_ti4 = simulate_system4(x_shifted_ti4, A_val, B_val)
y_orig_shifted_ti4 = np.zeros_like(t)
y_orig_shifted_ti4[k0_samples:] = y_orig_ti4[:-k0_samples]

print(f"Sistema 4 - Es invariante en el tiempo? {np.allclose(y_shifted_input_ti4, y_orig_shifted_ti4)}") # Esperado: True

plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.plot(t, y_sum_input_cont, 'b-', label='y[x1+x2]')
plt.plot(t, y_sum_outputs_cont, 'r--', label='y1+y2')
plt.title('Aditividad: y[x1+x2] (Azul) vs y1+y2 (Rojo)')
plt.legend()
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(t, y_ax_hom4, 'b-', label='y[a*x]')
plt.plot(t, y_a_times_y_x_hom4, 'r--', label='a*y[x]')
plt.title('Homogeneidad: y[a*x] (Azul) vs a*y[x] (Rojo)')
plt.legend()
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(t, y_shifted_input_ti4, 'b-', label='y[x(t-t0)]')
plt.plot(t, y_orig_shifted_ti4, 'r--', label='y[x(t)][t-t0]')
plt.title('Invariancia: y[x(t-t0)] (Azul) vs y[x(t)][t-t0] (Rojo)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Demostracin Analtica (Sistema 4):
Linealidad:
Aditividad: $A(x_1(t)+x_2(t)) + B \neq (Ax_1(t)+B) + (Ax_2(t)+B) = Ax_1(t)+Ax_2(t)+2B$, a menos que $B=0$.
Homogeneidad: $A(ax(t)) + B \neq a(Ax(t)+B) = aAx(t)+aB$, a menos que $B=0$ o $a=1$. Por lo tanto, no es lineal si $B \neq 0$.
Invariancia en el Tiempo: Un desplazamiento en la entrada $x(t-t_0)$ produce una salida $A x(t-t_0) + B$, que es idntica a la salida original $y(t)$ desplazada en $t_0$, $A x(t-t_0) + B$. S es invariante en el tiempo.
Conclusin para el Sistema 4: No es un SLIT (a menos que $B=0$, en cuyo caso s sera lineal y por ende un SLIT).
