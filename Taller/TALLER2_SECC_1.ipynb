## Comparación entre las formas de la serie de Fourier

La serie de Fourier es una herramienta fundamental en el análisis de señales, ya que permite representar cualquier señal periódica como una suma de funciones sinusoidales. Existen tres formas principales para expresar esta serie: **forma trigonométrica**, **forma exponencial** y **forma compacta**. Aunque todas describen lo mismo, difieren en la forma matemática de escribir la señal y en la interpretación de los coeficientes.

### 1. Serie de Fourier Trigonométrica

Esta forma representa la señal como una suma infinita de senos y cosenos. La fórmula general es:

$ x(t) = a_0 + \sum_{n=1}^{\infty} \left( a_n \cos(n \omega_0 t) + b_n \sin(n \omega_0 t) \right) $

- $a_0$ es el valor promedio (componente DC) de la señal.
- $a_n$ y $b_n$ son coeficientes reales que indican la amplitud de cada componente armónica.
- $\omega_0 = \frac{2\pi}{T}$ es la frecuencia fundamental.

Esta forma es fácil de interpretar porque cada término corresponde a una frecuencia específica, con una amplitud y una fase definida.

### 2. Serie de Fourier Exponencial

En esta versión, la señal se expresa usando exponenciales complejas, lo cual facilita muchos análisis matemáticos:

$ x(t) = \sum_{n=-\infty}^{\infty} C_n e^{j n \omega_0 t} $

- Los $C_n$ son coeficientes complejos que contienen toda la información de amplitud y fase.
- Esta forma es más compacta y general que la trigonométrica, y se usa mucho en teoría de sistemas y procesamiento de señales.

### 3. Forma Compacta

El término "forma compacta" suele referirse simplemente a la forma exponencial, ya que es una versión más condensada y algebraicamente manejable. Se basa en el uso de números complejos para representar la señal de forma más eficiente.

---

### Semejanzas

- Todas representan señales periódicas.
- Permiten descomponer una señal en sumas de funciones más simples (senos, cosenos o exponenciales).
- Los coeficientes de cada forma están relacionados y se pueden convertir de una forma a otra.

### Diferencias

| Característica          | Trigonométrica                            | Exponencial                                | Compacta                             |
|-------------------------|--------------------------------------------|---------------------------------------------|---------------------------------------|
| Funciones base          | Senos y cosenos                           | Exponenciales complejas                     | Equivalente a la exponencial          |
| Coeficientes            | Reales ($a_n$, $b_n$)                     | Complejos ($C_n$)                           | Complejos                             |
| Interpretación física   | Más intuitiva                             | Más técnica, útil para análisis matemático  | Similar a la exponencial              |
| Simetría                | Explícita                                 | Implícita ($C_{-n} = C_n^*$ para señales reales) | Idem exponencial                  |

## Transformada de Fourier, DTFT y DFT: semejanzas y diferencias

Después de entender las series de Fourier para señales periódicas, es importante conocer cómo se analiza una señal **no periódica** o **discreta**. Aquí es donde aparecen la **Transformada de Fourier (FT)**, la **Transformada de Fourier en tiempo discreto (DTFT)** y la **Transformada Discreta de Fourier (DFT)**. Aunque tienen nombres parecidos, cada una se usa en un contexto diferente.

### 1. Transformada de Fourier (FT)

Se aplica a **señales continuas en el tiempo** que pueden ser **no periódicas**. Convierte la señal del dominio del tiempo al dominio de la frecuencia continua.

$ X(f) = \int_{-\infty}^{\infty} x(t) e^{-j 2\pi f t} dt $

- $x(t)$ es la señal en el tiempo.
- $X(f)$ es la representación en frecuencia.
- El espectro resultante es **continuo**.

Se usa mucho en física, comunicaciones y procesamiento de señales analógicas.

### 2. Transformada de Fourier en Tiempo Discreto (DTFT)

Se aplica a señales **discretas en el tiempo pero infinitas**. También entrega un espectro **continuo**, pero periódico.

$ X(e^{j\omega}) = \sum_{n=-\infty}^{\infty} x[n] e^{-j \omega n} $

- $x[n]$ es la señal discreta.
- $X(e^{j\omega})$ es su representación en frecuencia, periódica en $\omega$ con periodo $2\pi$.

La DTFT es muy útil en teoría de procesamiento digital de señales.

### 3. Transformada Discreta de Fourier (DFT)

Se aplica a señales **discretas y de duración finita**. A diferencia de la DTFT, el resultado es también **discreto** (en frecuencia).

$ X[k] = \sum_{n=0}^{N-1} x[n] e^{-j \frac{2\pi}{N} kn} $

- $x[n]$ es la señal de entrada, con $N$ muestras.
- $X[k]$ representa las componentes de frecuencia discretas.
- Se implementa fácilmente en computador y es base para algoritmos como la FFT.

### Comparación general

| Aspecto                         | FT                           | DTFT                                | DFT                                |
|----------------------------------|-------------------------------|--------------------------------------|-------------------------------------|
| Tipo de señal                    | Continua, infinita           | Discreta, infinita                  | Discreta, finita                    |
| Resultado (espectro)            | Continuo                     | Continuo, pero periódico             | Discreto y periódico                |
| Aplicación                      | Señales analógicas           | Señales digitales teóricas          | Señales digitales reales (cómputo) |
| Implementación en computador    | No                           | No directamente                     | Sí                                  |

## ¿Qué es la FFT?

Aquí es donde entra el algoritmo **Fast Fourier Transform (FFT)**, que explicaremos en el siguiente punto.


| Concepto                            | Descripción                                                                                                                                     |
|-------------------------------------|-------------------------------------------------------------------------------------------------------------------------------------------------|
| **DFT (Transformada Discreta de Fourier)** | Calcula las componentes en frecuencia de una señal digital. Requiere muchas operaciones.                                                        |
| **Costo computacional DFT**         | $O(N^2)$, porque necesita $N$ sumas para cada uno de los $N$ resultados.                                                                        |
| **FFT (Transformada Rápida de Fourier)** | Es un algoritmo eficiente para calcular la DFT. Usa un enfoque de dividir y conquistar (por ejemplo, el algoritmo Cooley-Tukey).              |
| **Costo computacional FFT**         | $O(N \log_2 N)$, mucho más eficiente que la DFT directa.                                                                                        |
| **Ventaja principal de la FFT**     | Reduce significativamente el tiempo de cálculo, lo que la hace útil en procesamiento de señales, audio, imagen, etc.                            |
| **Método principal de la FFT**      | Divide la señal en partes (pares e impares), calcula la DFT de cada parte y luego combina los resultados.                                       |

## Transformadas de Fourier – Paso a Paso

### a)   $ x(t) = e^{-a|t|}, \; a > 0 $

$$
X(f) = \int_{-\infty}^{\infty} e^{-a|t|} e^{-j 2\pi f t} dt
$$

Partimos en dos integrales:

$$
X(f) = \int_{-\infty}^{0} e^{(a - j2\pi f)t} dt + \int_{0}^{\infty} e^{-(a + j2\pi f)t} dt
$$

Evaluamos ambas:

$$
X(f) = \frac{1}{a - j2\pi f} + \frac{1}{a + j2\pi f}
= \frac{2a}{a^2 + (2\pi f)^2}
$$

---

### b) $ x(t) = \cos(w_c t) $

Usamos la forma exponencial:

$$
\cos(w_c t) = \frac{1}{2}(e^{j w_c t} + e^{-j w_c t})
$$

Entonces:

$$
X(f) = \frac{1}{2} \left[ \delta\left(f - \frac{w_c}{2\pi}\right) + \delta\left(f + \frac{w_c}{2\pi}\right) \right]
$$

---

### c) $ x(t) = \sin(w_s t) $

Forma exponencial:

$$
\sin(w_s t) = \frac{1}{2j}(e^{j w_s t} - e^{-j w_s t})
$$

Transformada:

$$
X(f) = \frac{1}{2j} \left[ \delta\left(f - \frac{w_s}{2\pi}\right) - \delta\left(f + \frac{w_s}{2\pi}\right) \right]
$$

---

### d) $x(t) = f(t)\cos(w_c t) $

No se puede resolver sin conocer la forma exacta de \( f(t) \).  
En general, este caso implica una convolución en frecuencia, pero como **no se deben usar propiedades**, no se puede continuar directamente.

---

### e) $ x(t) = e^{-a|t|^2}, \; a > 0 $

Transformada de una gaussiana:

$$
X(f) = \int_{-\infty}^{\infty} e^{-a t^2} e^{-j 2\pi f t} dt
$$

Resultado conocido:

$$
X(f) = \sqrt{\frac{\pi}{a}} \, e^{-\frac{(\pi f)^2}{a}}
$$

---

### f) $ x(t) = A \, \text{rect}_d(t) $

La función \( \text{rect}_d(t) \) está definida en el intervalo \( [-d/2, d/2] \):

$$
X(f) = \int_{-d/2}^{d/2} A \, e^{-j 2\pi f t} dt
$$

Evaluamos:

$$
X(f) = A \cdot \frac{\sin(\pi f d)}{\pi f} = A d \cdot \text{sinc}(f d)
$$

Donde:

$$
\text{sinc}(x) = \frac{\sin(\pi x)}{\pi x}
$$
