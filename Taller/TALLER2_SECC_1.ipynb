## Comparación entre las formas de la serie de Fourier

La serie de Fourier es una herramienta fundamental en el análisis de señales, ya que permite representar cualquier señal periódica como una suma de funciones sinusoidales. Existen tres formas principales para expresar esta serie: **forma trigonométrica**, **forma exponencial** y **forma compacta**. Aunque todas describen lo mismo, difieren en la forma matemática de escribir la señal y en la interpretación de los coeficientes.

### 1. Serie de Fourier Trigonométrica

Esta forma representa la señal como una suma infinita de senos y cosenos. La fórmula general es:

$ x(t) = a_0 + \sum_{n=1}^{\infty} \left( a_n \cos(n \omega_0 t) + b_n \sin(n \omega_0 t) \right) $

- $a_0$ es el valor promedio (componente DC) de la señal.
- $a_n$ y $b_n$ son coeficientes reales que indican la amplitud de cada componente armónica.
- $\omega_0 = \frac{2\pi}{T}$ es la frecuencia fundamental.

Esta forma es fácil de interpretar porque cada término corresponde a una frecuencia específica, con una amplitud y una fase definida.

### 2. Serie de Fourier Exponencial

En esta versión, la señal se expresa usando exponenciales complejas, lo cual facilita muchos análisis matemáticos:

$ x(t) = \sum_{n=-\infty}^{\infty} C_n e^{j n \omega_0 t} $

- Los $C_n$ son coeficientes complejos que contienen toda la información de amplitud y fase.
- Esta forma es más compacta y general que la trigonométrica, y se usa mucho en teoría de sistemas y procesamiento de señales.

### 3. Forma Compacta

El término "forma compacta" suele referirse simplemente a la forma exponencial, ya que es una versión más condensada y algebraicamente manejable. Se basa en el uso de números complejos para representar la señal de forma más eficiente.

---

### Semejanzas

- Todas representan señales periódicas.
- Permiten descomponer una señal en sumas de funciones más simples (senos, cosenos o exponenciales).
- Los coeficientes de cada forma están relacionados y se pueden convertir de una forma a otra.

### Diferencias

| Característica          | Trigonométrica                            | Exponencial                                | Compacta                             |
|-------------------------|--------------------------------------------|---------------------------------------------|---------------------------------------|
| Funciones base          | Senos y cosenos                           | Exponenciales complejas                     | Equivalente a la exponencial          |
| Coeficientes            | Reales ($a_n$, $b_n$)                     | Complejos ($C_n$)                           | Complejos                             |
| Interpretación física   | Más intuitiva                             | Más técnica, útil para análisis matemático  | Similar a la exponencial              |
| Simetría                | Explícita                                 | Implícita ($C_{-n} = C_n^*$ para señales reales) | Idem exponencial                  |

## Transformada de Fourier, DTFT y DFT: semejanzas y diferencias

Después de entender las series de Fourier para señales periódicas, es importante conocer cómo se analiza una señal **no periódica** o **discreta**. Aquí es donde aparecen la **Transformada de Fourier (FT)**, la **Transformada de Fourier en tiempo discreto (DTFT)** y la **Transformada Discreta de Fourier (DFT)**. Aunque tienen nombres parecidos, cada una se usa en un contexto diferente.

### 1. Transformada de Fourier (FT)

Se aplica a **señales continuas en el tiempo** que pueden ser **no periódicas**. Convierte la señal del dominio del tiempo al dominio de la frecuencia continua.

$ X(f) = \int_{-\infty}^{\infty} x(t) e^{-j 2\pi f t} dt $

- $x(t)$ es la señal en el tiempo.
- $X(f)$ es la representación en frecuencia.
- El espectro resultante es **continuo**.

Se usa mucho en física, comunicaciones y procesamiento de señales analógicas.

### 2. Transformada de Fourier en Tiempo Discreto (DTFT)

Se aplica a señales **discretas en el tiempo pero infinitas**. También entrega un espectro **continuo**, pero periódico.

$ X(e^{j\omega}) = \sum_{n=-\infty}^{\infty} x[n] e^{-j \omega n} $

- $x[n]$ es la señal discreta.
- $X(e^{j\omega})$ es su representación en frecuencia, periódica en $\omega$ con periodo $2\pi$.

La DTFT es muy útil en teoría de procesamiento digital de señales.

### 3. Transformada Discreta de Fourier (DFT)

Se aplica a señales **discretas y de duración finita**. A diferencia de la DTFT, el resultado es también **discreto** (en frecuencia).

$ X[k] = \sum_{n=0}^{N-1} x[n] e^{-j \frac{2\pi}{N} kn} $

- $x[n]$ es la señal de entrada, con $N$ muestras.
- $X[k]$ representa las componentes de frecuencia discretas.
- Se implementa fácilmente en computador y es base para algoritmos como la FFT.

### Comparación general

| Aspecto                         | FT                           | DTFT                                | DFT                                |
|----------------------------------|-------------------------------|--------------------------------------|-------------------------------------|
| Tipo de señal                    | Continua, infinita           | Discreta, infinita                  | Discreta, finita                    |
| Resultado (espectro)            | Continuo                     | Continuo, pero periódico             | Discreto y periódico                |
| Aplicación                      | Señales analógicas           | Señales digitales teóricas          | Señales digitales reales (cómputo) |
| Implementación en computador    | No                           | No directamente                     | Sí                                  |

## ¿Qué es la FFT?

Aquí es donde entra el algoritmo **Fast Fourier Transform (FFT)**, que explicaremos en el siguiente punto.


| Concepto                            | Descripción                                                                                                                                     |
|-------------------------------------|-------------------------------------------------------------------------------------------------------------------------------------------------|
| **DFT (Transformada Discreta de Fourier)** | Calcula las componentes en frecuencia de una señal digital. Requiere muchas operaciones.                                                        |
| **Costo computacional DFT**         | $O(N^2)$, porque necesita $N$ sumas para cada uno de los $N$ resultados.                                                                        |
| **FFT (Transformada Rápida de Fourier)** | Es un algoritmo eficiente para calcular la DFT. Usa un enfoque de dividir y conquistar (por ejemplo, el algoritmo Cooley-Tukey).              |
| **Costo computacional FFT**         | $O(N \log_2 N)$, mucho más eficiente que la DFT directa.                                                                                        |
| **Ventaja principal de la FFT**     | Reduce significativamente el tiempo de cálculo, lo que la hace útil en procesamiento de señales, audio, imagen, etc.                            |
| **Método principal de la FFT**      | Divide la señal en partes (pares e impares), calcula la DFT de cada parte y luego combina los resultados.                                       |

## Transformadas de Fourier – Paso a Paso

### a)   $ x(t) = e^{-a|t|}, \; a > 0 $

$$
X(f) = \int_{-\infty}^{\infty} e^{-a|t|} e^{-j 2\pi f t} dt
$$

Partimos en dos integrales:

$$
X(f) = \int_{-\infty}^{0} e^{(a - j2\pi f)t} dt + \int_{0}^{\infty} e^{-(a + j2\pi f)t} dt
$$

Evaluamos ambas:

$$
X(f) = \frac{1}{a - j2\pi f} + \frac{1}{a + j2\pi f}
= \frac{2a}{a^2 + (2\pi f)^2}
$$

---

### b) $ x(t) = \cos(w_c t) $

Usamos la forma exponencial:

$$
\cos(w_c t) = \frac{1}{2}(e^{j w_c t} + e^{-j w_c t})
$$

Entonces:

$$
X(f) = \frac{1}{2} \left[ \delta\left(f - \frac{w_c}{2\pi}\right) + \delta\left(f + \frac{w_c}{2\pi}\right) \right]
$$

---

### c) $ x(t) = \sin(w_s t) $

Forma exponencial:

$$
\sin(w_s t) = \frac{1}{2j}(e^{j w_s t} - e^{-j w_s t})
$$

Transformada:

$$
X(f) = \frac{1}{2j} \left[ \delta\left(f - \frac{w_s}{2\pi}\right) - \delta\left(f + \frac{w_s}{2\pi}\right) \right]
$$

---

### d) $x(t) = f(t)\cos(w_c t) $

No se puede resolver sin conocer la forma exacta de \( f(t) \).  
En general, este caso implica una convolución en frecuencia, pero como **no se deben usar propiedades**, no se puede continuar directamente.

---

### e) $ x(t) = e^{-a|t|^2}, \; a > 0 $

Transformada de una gaussiana:

$$
X(f) = \int_{-\infty}^{\infty} e^{-a t^2} e^{-j 2\pi f t} dt
$$

Resultado conocido:

$$
X(f) = \sqrt{\frac{\pi}{a}} \, e^{-\frac{(\pi f)^2}{a}}
$$

---

### f) $ x(t) = A \, \text{rect}_d(t) $

La función \( \text{rect}_d(t) \) está definida en el intervalo \( [-d/2, d/2] \):

$$
X(f) = \int_{-d/2}^{d/2} A \, e^{-j 2\pi f t} dt
$$

Evaluamos:

$$
X(f) = A \cdot \frac{\sin(\pi f d)}{\pi f} = A d \cdot \text{sinc}(f d)
$$

Donde:

$$
\text{sinc}(x) = \frac{\sin(\pi x)}{\pi x}
$$

#Demostración Analítica (Sistema 1):
Linealidad: La presencia del término recursivo $-y[n-1]$ y su interacción con las entradas hacen que el sistema no sea lineal en su forma general, especialmente si las condiciones iniciales no son triviales o se asume el sistema en reposo de manera estricta. La superposición no se mantiene universalmente.
Invariancia en el Tiempo: El término $-y[n-1]$ también viola la invariancia temporal. Un retraso en la entrada afectaría la secuencia de $y[n-1]$ de forma que la salida retrasada no sería simplemente la salida original retrasada

Sistema 1: No es un SLIT.

```python

## 1. Sistema $y[n] = x[n]/3 + 2x[n-1] - y[n-1]$

### **Simulación en Python:**

import numpy as np
import matplotlib.pyplot as plt

def simulate_system1(x, initial_y=0):
    y = np.zeros_like(x, dtype=float)
    y_prev = initial_y
    for n in range(len(x)):
        x_n_minus_1 = x[n-1] if n > 0 else 0
        y[n] = x[n]/3 + 2 * x_n_minus_1 - y_prev
        y_prev = y[n]
    return y

N = 20 # Número de muestras

# --- Linealidad: Aditividad ---
x1 = np.zeros(N); x1[5:10] = 1
x2 = np.zeros(N); x2[12:15] = 0.5
y1 = simulate_system1(x1)
y2 = simulate_system1(x2)
y_sum_input = simulate_system1(x1 + x2)
y_sum_outputs = y1 + y2
print(f"Sistema 1 - ¿Es aditivo? {np.allclose(y_sum_input, y_sum_outputs)}") # Esperado: False

# --- Linealidad: Homogeneidad ---
a = 3
x_hom = np.zeros(N); x_hom[5:10] = 1
y_x_hom = simulate_system1(x_hom)
y_ax_hom = simulate_system1(a * x_hom)
print(f"Sistema 1 - ¿Es homogéneo? {np.allclose(a * y_x_hom, y_ax_hom)}") # Esperado: False

# --- Invariancia en el Tiempo ---
k0 = 5
x_orig_ti = np.zeros(N); x_orig_ti[5:10] = 1
x_shifted_ti = np.zeros(N)
if N - k0 > 0: x_shifted_ti[5+k0 : 10+k0] = 1
y_orig_ti = simulate_system1(x_orig_ti)
y_shifted_input_ti = simulate_system1(x_shifted_ti)
y_orig_shifted_ti = np.zeros(N)
if N - k0 > 0: y_orig_shifted_ti[k0:] = y_orig_ti[:-k0]
print(f"Sistema 1 - ¿Es invariante en el tiempo? {np.allclose(y_shifted_input_ti, y_orig_shifted_ti)}") # Esperado: False

plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.stem(range(N), y_sum_input, linefmt='b-', markerfmt='bo', basefmt=' ')
plt.stem(range(N), y_sum_outputs, linefmt='r--', markerfmt='rx', basefmt=' ')
plt.title('Aditividad: y[x1+x2] (Azul) vs y1+y2 (Rojo)')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.stem(range(N), y_ax_hom, linefmt='b-', markerfmt='bo', basefmt=' ')
plt.stem(range(N), a * y_x_hom, linefmt='r--', markerfmt='rx', basefmt=' ')
plt.title('Homogeneidad: y[a*x] (Azul) vs a*y[x] (Rojo)')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.stem(range(N), y_shifted_input_ti, linefmt='b-', markerfmt='bo', basefmt=' ')
plt.stem(range(N), y_orig_shifted_ti, linefmt='r--', markerfmt='rx', basefmt=' ')
plt.title('Invariancia: y[x[n-k0]] (Azul) vs y[x][n-k0] (Rojo)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

#2. Sistema $y[n] = \sum_{k=-\infty}^{n} x^2[k]$
Este sistema discreto involucra una operación no lineal ($x^2[k]$).

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_system2(x):
    y = np.zeros_like(x, dtype=float)
    cumulative_sum_sq = 0
    for n in range(len(x)):
        cumulative_sum_sq += x[n]**2
        y[n] = cumulative_sum_sq
    return y

N = 20

# --- Linealidad: Aditividad ---
x1 = np.zeros(N); x1[5] = 2
x2 = np.zeros(N); x2[6] = 3
y1 = simulate_system2(x1)
y2 = simulate_system2(x2)
y_sum_input = simulate_system2(x1 + x2)
y_sum_outputs = y1 + y2
print(f"Sistema 2 - ¿Es aditivo? {np.allclose(y_sum_input, y_sum_outputs)}") # Esperado: False

# --- Linealidad: Homogeneidad ---
a = 2
x_hom2 = np.zeros(N); x_hom2[5] = 2
y_x_hom2 = simulate_system2(x_hom2)
y_ax_hom2 = simulate_system2(a * x_hom2)
print(f"Sistema 2 - ¿Es homogéneo? {np.allclose(a * y_x_hom2, y_ax_hom2)}") # Esperado: False

# --- Invariancia en el Tiempo ---
k0 = 3
x_orig_ti2 = np.zeros(N); x_orig_ti2[5:8] = 1
x_shifted_ti2 = np.zeros(N)
if N - k0 > 0: x_shifted_ti2[5+k0 : 8+k0] = 1
y_orig_ti2 = simulate_system2(x_orig_ti2)
y_shifted_input_ti2 = simulate_system2(x_shifted_ti2)
y_orig_shifted_ti2 = np.zeros(N)
if N - k0 > 0: y_orig_shifted_ti2[k0:] = y_orig_ti2[:-k0]
print(f"Sistema 2 - ¿Es invariante en el tiempo? {np.allclose(y_shifted_input_ti2, y_orig_shifted_ti2)}") # Esperado: True

plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.stem(range(N), y_sum_input, linefmt='b-', markerfmt='bo', basefmt=' ')
plt.stem(range(N), y_sum_outputs, linefmt='r--', markerfmt='rx', basefmt=' ')
plt.title('Aditividad: y[x1+x2] (Azul) vs y1+y2 (Rojo)')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.stem(range(N), y_ax_hom2, linefmt='b-', markerfmt='bo', basefmt=' ')
plt.stem(range(N), a * y_x_hom2, linefmt='r--', markerfmt='rx', basefmt=' ')
plt.title('Homogeneidad: y[a*x] (Azul) vs a*y[x] (Rojo)')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.stem(range(N), y_shifted_input_ti2, linefmt='b-', markerfmt='bo', basefmt=' ')
plt.stem(range(N), y_orig_shifted_ti2, linefmt='r--', markerfmt='rx', basefmt=' ')
plt.title('Invariancia: y[x[n-k0]] (Azul) vs y[x][n-k0] (Rojo)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Demostración Analítica (Sistema 2):
Linealidad: La operación $x^2[k]$ rompe la linealidad.
Aditividad: $\sum (x_1+x_2)^2 \neq \sum x_1^2 + \sum x_2^2$ debido al término cruzado $2x_1x_2$.
Homogeneidad: $\sum (ax)^2 = a^2 \sum x^2 \neq a \sum x^2$ (a menos que $a=0$ o $a=1$). Por lo tanto, no es lineal.
Invariancia en el Tiempo: El operador de suma y la potencia no dependen del instante de tiempo absoluto $n$, solo del índice relativo $k$. Un desplazamiento en la entrada $x[n-k_0]$ resulta en un desplazamiento idéntico en la salida. Por lo tanto, sí es invariante en el tiempo.
Conclusión para el Sistema 2: No es un SLIT (por no ser lineal).

3. Sistema $y[n] = \text{median}(x[n-1], x[n], x[n+1])$
Este es un sistema discreto de filtrado de mediana con ventana de 3. La operación de mediana es no lineal.

Simulación en Python:

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_system3(x):
    y = np.zeros_like(x, dtype=float)
    x_padded = np.pad(x, (1, 1), 'constant', constant_values=0)
    for n in range(len(x)):
        window = [x_padded[n], x_padded[n+1], x_padded[n+2]]
        y[n] = np.median(window)
    return y

N = 10

# --- Linealidad: Aditividad ---
x1 = np.zeros(N); x1[3] = 10
x2 = np.zeros(N); x2[4] = 1
y1 = simulate_system3(x1)
y2 = simulate_system3(x2)
y_sum_input = simulate_system3(x1 + x2)
y_sum_outputs = y1 + y2
print(f"Sistema 3 - ¿Es aditivo? {np.allclose(y_sum_input, y_sum_outputs)}") # Esperado: False

# --- Linealidad: Homogeneidad ---
a = 2
x_hom3 = np.zeros(N); x_hom3[3:6] = [1, 5, 2]
y_x_hom3 = simulate_system3(x_hom3)
y_ax_hom3 = simulate_system3(a * x_hom3)
print(f"Sistema 3 - ¿Es homogéneo? {np.allclose(a * y_x_hom3, y_ax_hom3)}") # Esperado: True (para a>=0) o False si a<0 y elementos cruzan 0.

# --- Invariancia en el Tiempo ---
k0 = 2
x_orig_ti3 = np.zeros(N); x_orig_ti3[3:6] = [1, 5, 2]
x_shifted_ti3 = np.zeros(N)
if N - k0 > 0: x_shifted_ti3[3+k0 : 6+k0] = [1, 5, 2]
y_orig_ti3 = simulate_system3(x_orig_ti3)
y_shifted_input_ti3 = simulate_system3(x_shifted_ti3)
y_orig_shifted_ti3 = np.zeros(N)
if N - k0 > 0: y_orig_shifted_ti3[k0:] = y_orig_ti3[:-k0]
print(f"Sistema 3 - ¿Es invariante en el tiempo? {np.allclose(y_shifted_input_ti3, y_orig_shifted_ti3)}") # Esperado: True

plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.stem(range(N), y_sum_input, linefmt='b-', markerfmt='bo', basefmt=' ')
plt.stem(range(N), y_sum_outputs, linefmt='r--', markerfmt='rx', basefmt=' ')
plt.title('Aditividad: y[x1+x2] (Azul) vs y1+y2 (Rojo)')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.stem(range(N), y_ax_hom3, linefmt='b-', markerfmt='bo', basefmt=' ')
plt.stem(range(N), a * y_x_hom3, linefmt='r--', markerfmt='rx', basefmt=' ')
plt.title('Homogeneidad: y[a*x] (Azul) vs a*y[x] (Rojo)')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.stem(range(N), y_shifted_input_ti3, linefmt='b-', markerfmt='bo', basefmt=' ')
plt.stem(range(N), y_orig_shifted_ti3, linefmt='r--', markerfmt='rx', basefmt=' ')
plt.title('Invariancia: y[x[n-k0]] (Azul) vs y[x][n-k0] (Rojo)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Demostración Analítica (Sistema 3):
Linealidad:
Aditividad: La mediana de una suma no es la suma de las medianas (ej., median(0, 10, 5) = 5, pero median(0,10,0)+median(0,0,5) = 0+0=0). No es aditivo.
Homogeneidad: $\text{median}(a \cdot \text{ventana}) = a \cdot \text{median}(\text{ventana})$. Sí es homogéneo. Dado que falla la aditividad, no es lineal.
Invariancia en el Tiempo: La operación de mediana sobre una ventana es intrínsecamente independiente del tiempo absoluto. Un desplazamiento de la entrada resulta en el mismo desplazamiento de la salida. Sí es invariante en el tiempo.
Conclusión para el Sistema 3: No es un SLIT (por no ser lineal).


# 4. Sistema $y(t) = Ax(t) + B$; $A, B \in \mathbb{R}$

Este es un sistema continuo con un término constante aditivo.

Simulación en Python:

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_system4(x, A, B):
    return A * x + B

A_val = 2
B_val = 5 # Valor no nulo para la constante

# --- Linealidad: Aditividad ---
t = np.linspace(0, 10, 100)
x1_cont = np.sin(t)
x2_cont = np.cos(t)
y_sum_input_cont = simulate_system4(x1_cont + x2_cont, A_val, B_val)
y_sum_outputs_cont = simulate_system4(x1_cont, A_val, B_val) + simulate_system4(x2_cont, A_val, B_val)
print(f"Sistema 4 - ¿Es aditivo? {np.allclose(y_sum_input_cont, y_sum_outputs_cont)}") # Esperado: False si B != 0

# --- Linealidad: Homogeneidad ---
a_scale = 3
x_hom4 = np.sin(t)
y_ax_hom4 = simulate_system4(a_scale * x_hom4, A_val, B_val)
y_a_times_y_x_hom4 = a_scale * simulate_system4(x_hom4, A_val, B_val)
print(f"Sistema 4 - ¿Es homogéneo? {np.allclose(y_ax_hom4, y_a_times_y_x_hom4)}") # Esperado: False si B != 0 y a != 1

# --- Invariancia en el Tiempo ---
k0_time = 2 # Retraso en segundos
dt = t[1] - t[0]
k0_samples = int(k0_time / dt)

x_orig_ti4 = np.sin(t)
y_orig_ti4 = simulate_system4(x_orig_ti4, A_val, B_val)

x_shifted_ti4 = np.zeros_like(t)
x_shifted_ti4[k0_samples:] = x_orig_ti4[:-k0_samples]

y_shifted_input_ti4 = simulate_system4(x_shifted_ti4, A_val, B_val)
y_orig_shifted_ti4 = np.zeros_like(t)
y_orig_shifted_ti4[k0_samples:] = y_orig_ti4[:-k0_samples]

print(f"Sistema 4 - ¿Es invariante en el tiempo? {np.allclose(y_shifted_input_ti4, y_orig_shifted_ti4)}") # Esperado: True

plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.plot(t, y_sum_input_cont, 'b-', label='y[x1+x2]')
plt.plot(t, y_sum_outputs_cont, 'r--', label='y1+y2')
plt.title('Aditividad: y[x1+x2] (Azul) vs y1+y2 (Rojo)')
plt.legend()
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(t, y_ax_hom4, 'b-', label='y[a*x]')
plt.plot(t, y_a_times_y_x_hom4, 'r--', label='a*y[x]')
plt.title('Homogeneidad: y[a*x] (Azul) vs a*y[x] (Rojo)')
plt.legend()
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(t, y_shifted_input_ti4, 'b-', label='y[x(t-t0)]')
plt.plot(t, y_orig_shifted_ti4, 'r--', label='y[x(t)][t-t0]')
plt.title('Invariancia: y[x(t-t0)] (Azul) vs y[x(t)][t-t0] (Rojo)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Demostración Analítica (Sistema 4):
Linealidad:
Aditividad: $A(x_1(t)+x_2(t)) + B \neq (Ax_1(t)+B) + (Ax_2(t)+B) = Ax_1(t)+Ax_2(t)+2B$, a menos que $B=0$.
Homogeneidad: $A(ax(t)) + B \neq a(Ax(t)+B) = aAx(t)+aB$, a menos que $B=0$ o $a=1$. Por lo tanto, no es lineal si $B \neq 0$.
Invariancia en el Tiempo: Un desplazamiento en la entrada $x(t-t_0)$ produce una salida $A x(t-t_0) + B$, que es idéntica a la salida original $y(t)$ desplazada en $t_0$, $A x(t-t_0) + B$. Sí es invariante en el tiempo.
Conclusión para el Sistema 4: No es un SLIT (a menos que $B=0$, en cuyo caso sí sería lineal y por ende un SLIT).

# Solución al Ejercicio de Convolución Discreta y Respuesta al Escalón

Este documento aborda la resolución de un problema fundamental en el análisis de Señales y Sistemas, aplicando los conceptos de convolución discreta y la relación entre la respuesta al impulso y la respuesta al escalón. La metodología empleada combina el razonamiento conceptual con la validación computacional, haciendo uso de los recursos teóricos y prácticos presentados en los cuadernos `4_Convolucion_Discreta.ipynb` y `3_Respuesta_Escalon.ipynb`.

---

## Parte 1: Primer SLIT - Convolución Directa

La primera sección del problema requiere la determinación de la señal de salida $y[n]$ de un Sistema Lineal Invariante en el Tiempo (SLIT), dada una señal de entrada $x[n]$ y la respuesta al impulso $h[n]$ del sistema. La operación que rige esta relación es la convolución discreta.

### 1.1. Definición de las Señales

Se procede a la definición formal de las señales involucradas, indicando la posición del origen ($n=0$) para una correcta interpretación de los índices:

* **Señal de Entrada ($x[n]$):** $\{-15, 5, \mathbf{-3}^\uparrow, 0, 5, 7, -1\}$
    * $x[0] = -3$.
    * El rango de índices de $x[n]$ es de $n=-2$ a $n=4$.
    * La longitud de la señal $x[n]$ es de 7 muestras.
    * La posición de $x[0]$ dentro del array es el índice 2.

* **Respuesta al Impulso ($h[n]$):** $\{1, -2, \mathbf{0}^\uparrow, 1, -2\}$
    * $h[0] = 0$.
    * El rango de índices de $h[n]$ es de $n=-2$ a $n=2$.
    * La longitud de la señal $h[n]$ es de 5 muestras.
    * La posición de $h[0]$ dentro del array es el índice 2.

### 1.2. Cálculo de la Salida $y[n]$ mediante Convolución Discreta

La salida $y[n]$ se obtiene mediante la convolución discreta: $y[n] = x[n] * h[n]$.

#### Método Gráfico (Conceptualización)

Aunque no se presenta una representación gráfica explícita, el proceso de convolución se conceptualiza mediante la técnica de "volteo y desplazamiento" de una de las señales ($h[k]$ a $h[n-k]$), seguida de la multiplicación punto a punto con $x[k]$ para cada valor de $n$, y la posterior sumatoria de los productos. Este enfoque es fundamental para comprender la mecánica subyacente de la convolución.

La longitud de la señal de salida $y[n]$ es $L_x + L_h - 1$, donde $L_x$ y $L_h$ son las longitudes de $x[n]$ y $h[n]$ respectivamente. Para este caso, $7 + 5 - 1 = 11$ muestras. El índice inicial de $y[n]$ será la suma de los índices iniciales de $x[n]$ y $h[n]$, es decir, $(-2) + (-2) = -4$. Por lo tanto, la señal $y[n]$ abarca el rango de $n=-4$ a $n=6$.

#### Simulación en Python

Para la verificación computacional de la convolución, se emplea la función `numpy.convolve()`. Es esencial la correcta gestión de los offsets para mapear los índices de los arrays de NumPy a los índices reales $n$ de las señales discretas.


```python

import numpy as np

print("--- Parte 1: Convolución Directa ---")

# 1. Definición de las señales y sus offsets (posición de n=0 en el array NumPy).
#    El offset_x indica que x[0] se encuentra en la posición 2 del array `x`.
x = np.array([-15, 5, -3, 0, 5, 7, -1])
offset_x = 2

#    El offset_h indica que h[0] se encuentra en la posición 2 del array `h`.
h = np.array([1, -2, 0, 1, -2])
offset_h = 2

print(f"Señal de entrada x[n]: {x} (x[0] = {x[offset_x]})")
print(f"Respuesta al impulso h[n]: {h} (h[0] = {h[offset_h]})")

# 2. Realización de la operación de convolución.
y = np.convolve(x, h)

# 3. Determinación del offset para la señal de salida y[n].
#    El offset de la señal convolucionada es la suma de los offsets de las señales de entrada.
offset_y = offset_x + offset_h

# 4. Presentación de la señal de salida y[n] con sus índices reales.
print(f"\nSalida y[n] (representación en array NumPy): {y}")
print(f"El valor de y[n] en n=0 (y[0]) se localiza en el índice {offset_y} del array y, con valor {y[offset_y]}.\n")

print("Salida y[n] con correspondencia a sus índices 'n' reales:")
for i in range(len(y)):
    n_val = i - offset_y # Cálculo del índice 'n' real
    print(f"y[{n_val}] = {y[i]}")
```

#Parte 2: Segundo SLIT - Derivación de $h[n]$ y Convolución

La segunda parte del problema involucra un nuevo sistema cuya información se proporciona a través de su respuesta al escalón, $s[n]$, en lugar de su respuesta al impulso. Para proceder con la convolución, es necesario primero derivar la respuesta al impulso ($h_{new}[n]$) a partir de $s[n]$.

2.1. Definición de la Respuesta al Escalón $s[n]$
La señal de respuesta al escalón del segundo sistema se define como:

Respuesta al Escalón ($s[n]$): ${-1, 6, -10, \mathbf{3}^\uparrow, 1, -10, 2, 5}$
$s[0] = 3$.
El rango de índices de $s[n]$ es de $n=-3$ a $n=4$.
La longitud de la señal $s[n]$ es de 8 muestras.
La posición de $s[0]$ dentro del array es el índice 3.
2.2. Cálculo de la Respuesta al Impulso $h_{new}[n]$
Con base en la teoría de Sistemas Lineales e Invariantes en el Tiempo, la respuesta al impulso $h[n]$ puede obtenerse a partir de la respuesta al escalón $s[n]$ mediante la siguiente relación de diferencias:

$h_{new}[n] = s[n] - s[n-1]$

Esta operación computa la diferencia finita de la señal de respuesta al escalón, lo que corresponde a la respuesta del sistema a un impulso unitario.


```python
print("\n--- Parte 2: Derivación de h[n] y Convolución ---")

# 1. Definición de la respuesta al escalón s[n] y su offset.
s = np.array([-1, 6, -10, 3, 1, -10, 2, 5])
offset_s = 3 # s[0] se encuentra en la posición 3 del array.

print(f"Respuesta al escalón s[n]: {s} (s[0] = {s[offset_s]})")

# 2. Cálculo de la nueva respuesta al impulso h_new[n].
#    Para simular s[n-1], se crea un array `s_shifted` desplazando `s` un elemento a la derecha
#    y rellenando la primera posición con un cero (asumiendo s[n]=0 para n<n_min).
s_shifted = np.concatenate(([0.0], s[:-1]))

#    Se realiza la resta s[n] - s[n-1].
h_new = s - s_shifted

#    El offset de h_new[0] es el mismo que el de s[0].
offset_h_new = offset_s

print(f"\nNueva respuesta al impulso h_new[n] (representación en array NumPy): {h_new}")
print(f"El valor de h_new[n] en n=0 (h_new[0]) se localiza en el índice {offset_h_new} del array h_new, con valor {h_new[offset_h_new]}.\n")

print("Nueva respuesta al impulso h_new[n] con correspondencia a sus índices 'n' reales:")
for i in range(len(h_new)):
    n_val = i - offset_h_new
    print(f"h_new[{n_val}] = {h_new[i]}")
```

#2.3. Cálculo de la Nueva Salida $y_{new}[n]$ mediante Convolución
Con la respuesta al impulso $h_{new}[n]$ ya determinada para el segundo sistema, se procede a calcular su salida $y_{new}[n]$ utilizando la misma señal de entrada $x[n]$ original. Esto se realiza nuevamente mediante la operación de convolución discreta: $y_{new}[n] = x[n] * h_{new}[n]$.

```python
# La señal de entrada x[n] es la misma definida en la Parte 1.
# Las variables 'x' y 'offset_x' ya están disponibles.

# Realización de la convolución con la nueva respuesta al impulso h_new.
y_new = np.convolve(x, h_new)

# Determinación del offset para la nueva señal de salida y_new[n].
offset_y_new = offset_x + offset_h_new

# Presentación de la nueva señal de salida y_new[n] con sus índices reales.
print(f"\nSalida y_new[n] para el segundo sistema (representación en array NumPy): {y_new}")
print(f"El valor de y_new[n] en n=0 (y_new[0]) se localiza en el índice {offset_y_new} del array y_new, con valor {y_new[offset_y_new]}.\n")

print("Salida y_new[n] con correspondencia a sus índices 'n' reales:")
for i in range(len(y_new)):
    n_val = i - offset_y_new
    print(f"y_new[{n_val}] = {y_new[i]}")
```

# Solución al Ejercicio de Sistemas en Cascada con Señales Gaussianas

En esta ocasión, nos enfrentamos a un problema que involucra sistemas en cascada (conectados en serie), donde la señal de entrada y la respuesta al impulso de uno de los sistemas son funciones Gaussianas. El ejercicio nos pide determinar la salida del sistema combinado bajo dos configuraciones diferentes. Este análisis es fundamental para comprender cómo las propiedades de los sistemas individuales, especialmente la linealidad y la invarianza en el tiempo, afectan la señal resultante en arreglos complejos.

---

## Definición de Componentes del Sistema

Antes de abordar las configuraciones en serie, es crucial definir claramente los elementos que componen nuestro sistema combinado:

* **Señal de Entrada Gaussiana ($x(t)$):**
    $x(t) = e^{-at^2}$, con $a \in \mathbb{R}^+$ (una constante positiva).
    Esta es una función Gaussiana centrada en $t=0$, cuya anchura está inversamente relacionada con $a$.

* **Sistema A (No Lineal):**
    Este sistema tiene una relación entrada-salida definida por $y_A(t) = x^2(t)$.
    **Observación Importante:** La relación $y_A(t) = x^2(t)$ indica que el Sistema A es **no lineal**. La linealidad implica que la salida de una suma de entradas es la suma de las salidas individuales, y la salida de una entrada escalada es la salida escalada. En este caso, $(c x(t))^2 = c^2 x^2(t) \neq c (x^2(t))$, por lo que la propiedad de escalado no se cumple de manera lineal.

* **Sistema B (Lineal e Invariante en el Tiempo - LTI):**
    Este sistema es caracterizado por su respuesta al impulso $h_B(t) = B e^{-bt^2}$, con $B \in \mathbb{R}$ y $b \in \mathbb{R}^+$.
    Esta es también una función Gaussiana, lo que es común en el análisis de filtros y sistemas LTI.

---

## Parte a) Configuración en Serie: $x(t) \rightarrow h_B(t) \rightarrow y_A(t) \rightarrow y(t)$

En esta configuración, la señal de entrada $x(t)$ primero pasa por el Sistema B (LTI con respuesta al impulso $h_B(t)$), y la salida de este, digamos $z(t)$, luego se convierte en la entrada del Sistema A, cuya salida final es $y(t)$.

El diagrama de bloques es: $x(t) \longrightarrow \boxed{\text{Sistema B}} \longrightarrow z(t) \longrightarrow \boxed{\text{Sistema A}} \longrightarrow y(t)$

### Salida del Sistema B ($z(t)$)

Dado que el Sistema B es LTI, su salida $z(t)$ se obtiene mediante la **convolución** de la entrada $x(t)$ con su respuesta al impulso $h_B(t)$:
$z(t) = x(t) * h_B(t)$

Para la convolución de dos funciones Gaussianas, existe una propiedad importante: la convolución de dos Gaussianas es otra Gaussiana.
Si $f_1(t) = C_1 e^{-a_1 t^2}$ y $f_2(t) = C_2 e^{-a_2 t^2}$, entonces $f_1(t) * f_2(t) = C_1 C_2 \sqrt{\frac{\pi}{a_1+a_2}} e^{-\frac{a_1 a_2}{a_1+a_2} t^2}$.

Aplicando esto a $x(t) = e^{-at^2}$ (donde $C_1=1, a_1=a$) y $h_B(t) = B e^{-bt^2}$ (donde $C_2=B, a_2=b$):

$z(t) = (e^{-at^2}) * (B e^{-bt^2})$
$z(t) = B \sqrt{\frac{\pi}{a+b}} e^{-\frac{ab}{a+b} t^2}$

### Salida Final del Sistema A ($y(t)$)

La salida $z(t)$ del Sistema B se convierte en la entrada del Sistema A. El Sistema A opera aplicando la relación $y_A(t) = x_{entrada\_A}^2(t)$. En este caso, $x_{entrada\_A}(t) = z(t)$.

Por lo tanto:
$y(t) = (z(t))^2$
$y(t) = \left( B \sqrt{\frac{\pi}{a+b}} e^{-\frac{ab}{a+b} t^2} \right)^2$
$y(t) = B^2 \left( \sqrt{\frac{\pi}{a+b}} \right)^2 \left( e^{-\frac{ab}{a+b} t^2} \right)^2$
$y(t) = B^2 \frac{\pi}{a+b} e^{-2 \frac{ab}{a+b} t^2}$

Esta es la salida final $y(t)$ para la configuración en serie a).

---

## Parte b) Configuración en Serie: $x(t) \rightarrow y_A(t) \rightarrow h_B(t) \rightarrow y(t)$

En esta segunda configuración, la señal de entrada $x(t)$ primero pasa por el Sistema A (no lineal), y la salida de este, digamos $w(t)$, luego se convierte en la entrada del Sistema B (LTI con respuesta al impulso $h_B(t)$), cuya salida final es $y(t)$.

El diagrama de bloques es: $x(t) \longrightarrow \boxed{\text{Sistema A}} \longrightarrow w(t) \longrightarrow \boxed{\text{Sistema B}} \longrightarrow y(t)$

### Salida del Sistema A ($w(t)$)

La entrada al Sistema A es $x(t)$, y su relación de entrada-salida es $y_A(t) = x^2(t)$. Entonces, la salida del Sistema A, $w(t)$, es:
$w(t) = (x(t))^2$
$w(t) = (e^{-at^2})^2$
$w(t) = e^{-2at^2}$

### Salida Final del Sistema B ($y(t)$)

La salida $w(t)$ del Sistema A se convierte en la entrada del Sistema B. Dado que el Sistema B es LTI, su salida $y(t)$ se obtiene mediante la **convolución** de su entrada $w(t)$ con su respuesta al impulso $h_B(t)$:
$y(t) = w(t) * h_B(t)$

Aplicando nuevamente la propiedad de convolución de dos Gaussianas:
$w(t) = e^{-2at^2}$ (donde $C_1=1, a_1=2a$)
$h_B(t) = B e^{-bt^2}$ (donde $C_2=B, a_2=b$)

$y(t) = (e^{-2at^2}) * (B e^{-bt^2})$
$y(t) = B \sqrt{\frac{\pi}{2a+b}} e^{-\frac{2ab}{2a+b} t^2}$

Esta es la salida final $y(t)$ para la configuración en serie b).



Hemos encontrado las expresiones para la salida $y(t)$ en ambas configuraciones de los sistemas en cascada. Es crucial notar cómo el orden de los sistemas afecta la salida final cuando uno de los sistemas es no lineal. La convolución, al ser una operación lineal, se ve precedida o seguida por una operación no lineal (el cuadrado de la señal), lo que cambia significativamente la forma final de la expresión. Esto subraya la importancia de la linealidad y la invarianza en el tiempo en el análisis de sistemas.

¡Claro que sí! Con mucho gusto te ayudo a demostrar estas propiedades de la Transformada de Laplace. La clave es usar siempre la definición original de la transformada y un poco de álgebra. Lo haré paso a paso para que sea fácil de seguir.

La definición de la Transformada de Laplace que usaremos es:
$$ \mathcal{L}\{f(t)\} = F(s) = \int_0^\infty f(t) e^{-st} dt $$

---

### i) Propiedad de Desplazamiento en el Tiempo: $\mathcal{L}\{x(t-t_o)\} = e^{-st_o} X(s)$

Esta propiedad nos dice qué pasa cuando retrasamos una función en el tiempo. Para esta demostración, asumimos que la función $x(t)$ es cero para $t < 0$ y que el desplazamiento $t_o$ es positivo.

1.  **Empezamos con la definición:**
    Aplicamos la definición de la transformada a nuestra función desplazada, $x(t-t_o)$:
    $$ \mathcal{L}\{x(t-t_o)\} = \int_0^\infty x(t-t_o) e^{-st} dt $$

2.  **Hacemos un cambio de variable:**
    Para simplificar la integral, vamos a cambiar la variable. Llamemos a una nueva variable $\tau$ (tau):
    * Sea $\tau = t - t_o$
    * Esto significa que $t = \tau + t_o$
    * Y el diferencial $dt$ se convierte en $d\tau$.

3.  **Ajustamos los límites de la integral:**
    Como cambiamos la variable, tenemos que ver cómo cambian los límites de la integral:
    * Cuando $t = 0$, $\tau = 0 - t_o = -t_o$.
    * Cuando $t \to \infty$, $\tau \to \infty$.
    La integral ahora se ve así:
    $$ \int_{-t_o}^\infty x(\tau) e^{-s(\tau+t_o)} d\tau $$

4.  **Usamos la propiedad de causalidad:**
    Por definición de la transformada unilateral, la función $x(\tau)$ es $0$ para cualquier valor de $\tau$ negativo. Como empezamos la integral en $-t_o$, la función será cero desde $-t_o$ hasta $0$. Así que podemos empezar a integrar desde $0$ sin cambiar el resultado.
    $$ \int_0^\infty x(\tau) e^{-s(\tau+t_o)} d\tau $$

5.  **Simplificamos y reordenamos:**
    Ahora, separamos la exponencial:
    $$ \int_0^\infty x(\tau) e^{-s\tau} e^{-st_o} d\tau $$
    El término $e^{-st_o}$ es una constante con respecto a $\tau$, así que puede salir de la integral:
    $$ e^{-st_o} \int_0^\infty x(\tau) e^{-s\tau} d\tau $$

6.  **Reconocemos la transformada original:**
    ¡Mira bien la integral que quedó! Es exactamente la definición de la Transformada de Laplace de $x(t)$, que es $X(s)$.
    $$ \int_0^\infty x(\tau) e^{-s\tau} d\tau = X(s) $$

7.  **Resultado final:**
    Juntando todo, llegamos a la propiedad que queríamos demostrar:
    $$ \mathcal{L}\{x(t-t_o)\} = e^{-st_o} X(s) $$

---

### ii) Propiedad de Escalado en el Tiempo: $\mathcal{L}\{x(at)\} = \frac{1}{|a|} X(s/a)$

Esta propiedad nos muestra cómo se afecta la transformada si "aceleramos" o "ralentizamos" el tiempo en la función. Para la demostración, consideremos el caso más común donde $a > 0$.

1.  **Empezamos con la definición:**
    $$ \mathcal{L}\{x(at)\} = \int_0^\infty x(at) e^{-st} dt $$

2.  **Hacemos un cambio de variable:**
    Llamemos a nuestra nueva variable $\tau$:
    * Sea $\tau = at$
    * Entonces $t = \tau/a$
    * Y el diferencial $dt = (1/a) d\tau$.

3.  **Ajustamos los límites de la integral:**
    * Cuando $t=0$, $\tau = a \cdot 0 = 0$.
    * Cuando $t \to \infty$, $\tau \to \infty$ (porque asumimos $a>0$).
    La integral se transforma en:
    $$ \int_0^\infty x(\tau) e^{-s(\tau/a)} \frac{1}{a} d\tau $$

4.  **Reordenamos la expresión:**
    El término $1/a$ es una constante, así que lo sacamos de la integral. También reorganizamos el exponente:
    $$ \frac{1}{a} \int_0^\infty x(\tau) e^{-(s/a)\tau} d\tau $$

5.  **Reconocemos la transformada original:**
    La integral que nos queda es muy parecida a la definición de $X(s)$, pero en todos los lugares donde debería haber una 's', ahora tenemos '(s/a)'. Esto significa que es la transformada $X$ evaluada en $(s/a)$.
    $$ \int_0^\infty x(\tau) e^{-(s/a)\tau} d\tau = X(s/a) $$

6.  **Resultado final:**
    Por lo tanto, obtenemos:
    $$ \mathcal{L}\{x(at)\} = \frac{1}{a} X(s/a) $$
    (Como asumimos $a>0$, entonces $a = |a|$, lo que coincide con la fórmula general).

### iii) Propiedad de Derivación: $\mathcal{L}\{\frac{dx(t)}{dt}\} = sX(s)$

Esta es una propiedad muy importante. La fórmula que te dieron es un caso especial donde se asume que la condición inicial $x(0) = 0$. Vamos a demostrar la fórmula completa y luego verás por qué se simplifica.

1.  **Empezamos con la definición:**
    $$ \mathcal{L}\left\{\frac{dx(t)}{dt}\right\} = \int_0^\infty \frac{dx(t)}{dt} e^{-st} dt $$

2.  **Usamos Integración por Partes:**
    Esta integral es un caso clásico para resolver por partes. Recordemos la fórmula: $\int u dv = uv - \int v du$.
    * Elijamos $u = e^{-st}$  y  $dv = \frac{dx(t)}{dt} dt (= dx)$.
    * Entonces, al derivar $u$ e integrar $dv$, obtenemos: $du = -se^{-st} dt$ y $v = x(t)$.

3.  **Aplicamos la fórmula:**
    Sustituyendo en la fórmula de integración por partes:
    $$ \mathcal{L}\left\{\frac{dx(t)}{dt}\right\} = [e^{-st} x(t)]_0^\infty - \int_0^\infty x(t) (-se^{-st}) dt $$

4.  **Evaluamos la primera parte:**
    $$ [e^{-st} x(t)]_0^\infty = \lim_{t \to \infty} (e^{-st} x(t)) - (e^{-s \cdot 0} x(0)) $$
    Para que la transformada de Laplace exista, se requiere que $\lim_{t \to \infty} e^{-st} x(t) = 0$. El segundo término es $e^0 x(0) = x(0)$.
    Así que esta parte nos da: $0 - x(0) = -x(0)$.

5.  **Evaluamos la segunda parte (la integral):**
    $$ - \int_0^\infty x(t) (-se^{-st}) dt $$
    Podemos sacar las constantes $(-s)$ de la integral, que con el signo de menos de afuera se cancelan:
    $$ s \int_0^\infty x(t) e^{-st} dt $$
    La integral que queda es, por definición, $X(s)$. Así que esta parte es $sX(s)$.

6.  **Juntamos todo:**
    La demostración completa es:
    $$ \mathcal{L}\left\{\frac{dx(t)}{dt}\right\} = sX(s) - x(0) $$
    Como puedes ver, si la condición inicial es cero, es decir, **si $x(0)=0$**, la fórmula se simplifica a la que te pidieron demostrar:
    $$ \mathcal{L}\left\{\frac{dx(t)}{dt}\right\} = sX(s) $$

### iv) Propiedad de Convolución: $\mathcal{L}\{x(t) * y(t)\} = X(s)Y(s)$

La convolución puede parecer intimidante, pero su transformada es muy elegante.

1.  **Recordemos la definición de convolución:**
    La convolución de dos funciones causales $x(t)$ e $y(t)$ es:
    $$ x(t) * y(t) = \int_0^t x(\tau) y(t-\tau) d\tau $$

2.  **Aplicamos la definición de la transformada de Laplace:**
    $$ \mathcal{L}\{x(t) * y(t)\} = \int_0^\infty \left[ \int_0^t x(\tau) y(t-\tau) d\tau \right] e^{-st} dt $$

3.  **Cambiamos el orden de integración:**
    Este es el paso clave. Estamos integrando en una región del plano $t-\tau$. Cambiar el orden de las integrales nos simplificará la vida. La región de integración es $0 \le \tau \le t$ y $0 \le t < \infty$. Si lo reordenamos, integramos primero respecto a $t$ desde $\tau$ hasta $\infty$, y luego respecto a $\tau$ desde $0$ hasta $\infty$.
    $$ \int_0^\infty \int_\tau^\infty \left[ x(\tau) y(t-\tau) d\tau \right] e^{-st} dt = \int_0^\infty x(\tau) \left[ \int_\tau^\infty y(t-\tau) e^{-st} dt \right] d\tau $$

4.  **Resolvemos la integral de adentro:**
    Fijémonos en la integral interna: $\int_\tau^\infty y(t-\tau) e^{-st} dt$.
    Hacemos un cambio de variable. Sea $u = t - \tau$.
    * Entonces $t = u + \tau$ y $dt = du$.
    * Los nuevos límites son: si $t=\tau \implies u=0$, y si $t \to \infty \implies u \to \infty$.
    La integral interna se convierte en:
    $$ \int_0^\infty y(u) e^{-s(u+\tau)} du = \int_0^\infty y(u) e^{-su} e^{-s\tau} du $$
    Sacamos $e^{-s\tau}$ (es constante respecto a $u$):
    $$ e^{-s\tau} \int_0^\infty y(u) e^{-su} du $$
    La integral es la definición de la transformada de $y(t)$, es decir, $Y(s)$. Así que la integral interna es $e^{-s\tau} Y(s)$.

5.  **Sustituimos de vuelta en la integral de afuera:**
    Ahora ponemos este resultado de nuevo en la expresión del paso 3:
    $$ \mathcal{L}\{x(t) * y(t)\} = \int_0^\infty x(\tau) [e^{-s\tau} Y(s)] d\tau $$

6.  **Resultado final:**
    $Y(s)$ es una constante con respecto a $\tau$, así que también puede salir de la integral:
    $$ Y(s) \int_0^\infty x(\tau) e^{-s\tau} d\tau $$
    Y la integral que queda es, por supuesto, la definición de $X(s)$. Por lo tanto, llegamos al resultado final:
    $$ \mathcal{L}\{x(t) * y(t)\} = X(s)Y(s) $$

### 1. Circuito RLC Serie

**Objetivo:** Encontrar la función de transferencia $H(s) = \frac{V_C(s)}{V_{in}(s)}$, donde la salida es la tensión en el capacitor.

**Análisis del Circuito:**
1.  Imaginemos una fuente de tensión $v_{in}(t)$ conectada en serie a una resistencia $R$, un inductor $L$ y un capacitor $C$.
2.  Aplicamos la **Ley de Voltajes de Kirchhoff (LVK)** alrededor de la malla. La suma de las caídas de tensión en los componentes es igual a la tensión de la fuente:
    $$v_{in}(t) = v_R(t) + v_L(t) + v_C(t)$$
3.  Expresamos cada tensión en términos de la corriente de la malla, $i(t)$:
    * $v_R(t) = R \cdot i(t)$
    * $v_L(t) = L \frac{di(t)}{dt}$
    * $v_C(t)$ es nuestra salida.

4.  La relación clave entre la corriente y el voltaje del capacitor es:
    $$i(t) = C \frac{dv_C(t)}{dt}$$
5.  Ahora, sustituimos $i(t)$ en las expresiones de voltaje para que todo quede en función de nuestra salida $v_C(t)$:
    * $v_R(t) = R \left( C \frac{dv_C(t)}{dt} \right) = RC \frac{dv_C(t)}{dt}$
    * Para $v_L(t)$, primero necesitamos la derivada de la corriente: $\frac{di(t)}{dt} = C \frac{d^2v_C(t)}{dt^2}$.
    * $v_L(t) = L \left( C \frac{d^2v_C(t)}{dt^2} \right) = LC \frac{d^2v_C(t)}{dt^2}$

6.  Sustituimos todo en la ecuación de la LVK:
    $$v_{in}(t) = \left( RC \frac{dv_C(t)}{dt} \right) + \left( LC \frac{d^2v_C(t)}{dt^2} \right) + v_C(t)$$
7.  Reordenando los términos, obtenemos la ecuación diferencial del sistema:
    $$LC \frac{d^2v_C(t)}{dt^2} + RC \frac{dv_C(t)}{dt} + v_C(t) = v_{in}(t)$$
    Si llamamos a la entrada $x(t) = v_{in}(t)$ y a la salida $y(t) = v_C(t)$, la ecuación es idéntica a la que tú proporcionaste. ¡Excelente!

**Cálculo de la Función de Transferencia:**

1.  Aplicamos la Transformada de Laplace a la ecuación diferencial (con condiciones iniciales cero):
    $\mathcal{L}\left\{ LC \frac{d^2y(t)}{dt^2} + RC \frac{dy(t)}{dt} + y(t) \right\} = \mathcal{L}\{x(t)\}$
    $LC \cdot s^2Y(s) + RC \cdot sY(s) + Y(s) = X(s)$
2.  Factorizamos $Y(s)$:
    $Y(s) (LCs^2 + RCs + 1) = X(s)$
3.  Despejamos la relación $Y(s)/X(s)$:
    $$H(s) = \frac{Y(s)}{X(s)} = \frac{V_C(s)}{V_{in}(s)} = \frac{1}{LCs^2 + RCs + 1}$$

Esta es la función de transferencia para el circuito RLC serie con salida en el capacitor.

### 2. Circuito RLC Paralelo

**Objetivo:** Encontrar la función de transferencia $H(s) = \frac{I_L(s)}{I_{in}(s)}$, donde la salida es la corriente en el inductor.

**Análisis del Circuito:**
1.  Imaginemos una fuente de corriente $i_{in}(t)$ que alimenta en paralelo a una resistencia $R$, un inductor $L$ y un capacitor $C$.
2.  Aplicamos la **Ley de Corrientes de Kirchhoff (LCK)** en el nodo superior. La corriente que entra es igual a la suma de las que salen:
    $$i_{in}(t) = i_R(t) + i_L(t) + i_C(t)$$
3.  La tensión $v(t)$ es la misma para todos los componentes. Expresamos cada corriente en términos de esta tensión:

    * $i_L(t)$ es nuestra salida.
    * $i_C(t) = C \frac{dv(t)}{dt}$

4.  La relación clave entre el voltaje y la corriente del inductor es:
    $$v(t) = L \frac{di_L(t)}{dt}$$
5.  Sustituimos $v(t)$ en las otras expresiones de corriente para que todo quede en función de nuestra salida $i_L(t)$:
    * $i_R(t) = \frac{1}{R} \left( L \frac{di_L(t)}{dt} \right) = \frac{L}{R} \frac{di_L(t)}{dt}$
    * Para $i_C(t)$, primero necesitamos la derivada del voltaje: $\frac{dv(t)}{dt} = L \frac{d^2i_L(t)}{dt^2}$.
    * $i_C(t) = C \left( L \frac{d^2i_L(t)}{dt^2} \right) = LC \frac{d^2i_L(t)}{dt^2}$

6.  Sustituimos todo en la ecuación de la LCK:
    $$i_{in}(t) = \left( \frac{L}{R} \frac{di_L(t)}{dt} \right) + i_L(t) + \left( LC \frac{d^2i_L(t)}{dt^2} \right)$$
7.  Reordenando, obtenemos la ecuación diferencial (nota que es diferente a la de la imagen):
    $$LC \frac{d^2i_L(t)}{dt^2} + \frac{L}{R} \frac{di_L(t)}{dt} + i_L(t) = i_{in}(t)$$

**Cálculo de la Función de Transferencia:**

1.  Aplicamos la Transformada de Laplace a la ecuación (con condiciones iniciales cero):

    $LC \cdot s^2Y(s) + \frac{L}{R} \cdot sY(s) + Y(s) = X(s)$
2.  Factorizamos $Y(s)$:
    $Y(s) \left(LCs^2 + \frac{L}{R}s + 1\right) = X(s)$
3.  Despejamos la relación $Y(s)/X(s)$:
    $$H(s) = \frac{Y(s)}{X(s)} = \frac{I_L(s)}{I_{in}(s)} = \frac{1}{LCs^2 + \frac{L}{R}s + 1}$$

    $$H(s) = \frac{\frac{1}{LC}}{s^2 + \frac{1}{RC}s + \frac{1}{LC}}$$

Esta es la función de transferencia para el circuito RLC paralelo con salida en la corriente del inductor.

## Parte 1: Demostración de Propiedades de la Transformada de Laplace



La definición de la Transformada de Laplace que usaremos es:
$$ \mathcal{L}\{f(t)\} = F(s) = \int_0^\infty f(t) e^{-st} dt $$

---

### i) Propiedad de Desplazamiento en el Tiempo: $\mathcal{L}\{x(t-t_o)\} = e^{-st_o} X(s)$

Esta propiedad nos dice qué pasa cuando retrasamos una función en el tiempo. Para esta demostración, asumimos que la función $x(t)$ es cero para $t < 0$ y que el desplazamiento $t_o$ es positivo.

1.  **Empezamos con la definición:**
    Aplicamos la definición de la transformada a nuestra función desplazada, $x(t-t_o)$:
    $$ \mathcal{L}\{x(t-t_o)\} = \int_0^\infty x(t-t_o) e^{-st} dt $$

2.  **Hacemos un cambio de variable:**
    Para simplificar la integral, vamos a cambiar la variable. Llamemos a una nueva variable $\tau$ (tau):
    * Sea $\tau = t - t_o$
    * Esto significa que $t = \tau + t_o$
    * Y el diferencial $dt$ se convierte en $d\tau$.

3.  **Ajustamos los límites de la integral:**
    Como cambiamos la variable, tenemos que ver cómo cambian los límites de la integral:
    * Cuando $t = 0$, $\tau = 0 - t_o = -t_o$.
    * Cuando $t \to \infty$, $\tau \to \infty$.
    La integral ahora se ve así:
    $$ \int_{-t_o}^\infty x(\tau) e^{-s(\tau+t_o)} d\tau $$

4.  **Usamos la propiedad de causalidad:**
    Por definición de la transformada unilateral, la función $x(\tau)$ es $0$ para cualquier valor de $\tau$ negativo. Como empezamos la integral en $-t_o$, la función será cero desde $-t_o$ hasta $0$. Así que podemos empezar a integrar desde $0$ sin cambiar el resultado.
    $$ \int_0^\infty x(\tau) e^{-s(\tau+t_o)} d\tau $$

5.  **Simplificamos y reordenamos:**
    Ahora, separamos la exponencial:
    $$ \int_0^\infty x(\tau) e^{-s\tau} e^{-st_o} d\tau $$
    El término $e^{-st_o}$ es una constante con respecto a $\tau$, así que puede salir de la integral:
    $$ e^{-st_o} \int_0^\infty x(\tau) e^{-s\tau} d\tau $$

6.  **Reconocemos la transformada original:**
    ¡Mira bien la integral que quedó! Es exactamente la definición de la Transformada de Laplace de $x(t)$, que es $X(s)$.
    $$ \int_0^\infty x(\tau) e^{-s\tau} d\tau = X(s) $$

7.  **Resultado final:**
    Juntando todo, llegamos a la propiedad que queríamos demostrar:
    $$ \mathcal{L}\{x(t-t_o)\} = e^{-st_o} X(s) $$

---

### ii) Propiedad de Escalado en el Tiempo: $\mathcal{L}\{x(at)\} = \frac{1}{|a|} X(s/a)$

Esta propiedad nos muestra cómo se afecta la transformada si "aceleramos" o "ralentizamos" el tiempo en la función. Para la demostración, consideremos el caso más común donde $a > 0$.

1.  **Empezamos con la definición:**
    $$ \mathcal{L}\{x(at)\} = \int_0^\infty x(at) e^{-st} dt $$

2.  **Hacemos un cambio de variable:**
    Llamemos a nuestra nueva variable $\tau$:
    * Sea $\tau = at$
    * Entonces $t = \tau/a$
    * Y el diferencial $dt = (1/a) d\tau$.

3.  **Ajustamos los límites de la integral:**
    * Cuando $t=0$, $\tau = a \cdot 0 = 0$.
    * Cuando $t \to \infty$, $\tau \to \infty$ (porque asumimos $a>0$).
    La integral se transforma en:
    $$ \int_0^\infty x(\tau) e^{-s(\tau/a)} \frac{1}{a} d\tau $$

4.  **Reordenamos la expresión:**
    El término $1/a$ es una constante, así que lo sacamos de la integral. También reorganizamos el exponente:
    $$ \frac{1}{a} \int_0^\infty x(\tau) e^{-(s/a)\tau} d\tau $$

5.  **Reconocemos la transformada original:**
    La integral que nos queda es muy parecida a la definición de $X(s)$, pero en todos los lugares donde debería haber una 's', ahora tenemos '(s/a)'. Esto significa que es la transformada $X$ evaluada en $(s/a)$.
    $$ \int_0^\infty x(\tau) e^{-(s/a)\tau} d\tau = X(s/a) $$

6.  **Resultado final:**
    Por lo tanto, obtenemos:
    $$ \mathcal{L}\{x(at)\} = \frac{1}{a} X(s/a) $$
    (Como asumimos $a>0$, entonces $a = |a|$, lo que coincide con la fórmula general).

---

### iii) Propiedad de Derivación: $\mathcal{L}\{\frac{dx(t)}{dt}\} = sX(s)$

Esta es una propiedad muy importante. La fórmula que te dieron es un caso especial donde se asume que la condición inicial $x(0) = 0$. Vamos a demostrar la fórmula completa y luego verás por qué se simplifica.

1.  **Empezamos con la definición:**
    $$ \mathcal{L}\left\{\frac{dx(t)}{dt}\right\} = \int_0^\infty \frac{dx(t)}{dt} e^{-st} dt $$

2.  **Usamos Integración por Partes:**
    Esta integral es un caso clásico para resolver por partes. Recordemos la fórmula: $\int u dv = uv - \int v du$.
    * Elijamos $u = e^{-st}$  y  $dv = \frac{dx(t)}{dt} dt (= dx)$.
    * Entonces, al derivar $u$ e integrar $dv$, obtenemos: $du = -se^{-st} dt$ y $v = x(t)$.

3.  **Aplicamos la fórmula:**
    Sustituyendo en la fórmula de integración por partes:
    $$ \mathcal{L}\left\{\frac{dx(t)}{dt}\right\} = [e^{-st} x(t)]_0^\infty - \int_0^\infty x(t) (-se^{-st}) dt $$

4.  **Evaluamos la primera parte:**
    $$ [e^{-st} x(t)]_0^\infty = \lim_{t \to \infty} (e^{-st} x(t)) - (e^{-s \cdot 0} x(0)) $$
    Para que la transformada de Laplace exista, se requiere que $\lim_{t \to \infty} e^{-st} x(t) = 0$. El segundo término es $e^0 x(0) = x(0)$.
    Así que esta parte nos da: $0 - x(0) = -x(0)$.

5.  **Evaluamos la segunda parte (la integral):**
    $$ - \int_0^\infty x(t) (-se^{-st}) dt $$
    Podemos sacar las constantes $(-s)$ de la integral, que con el signo de menos de afuera se cancelan:
    $$ s \int_0^\infty x(t) e^{-st} dt $$
    La integral que queda es, por definición, $X(s)$. Así que esta parte es $sX(s)$.

6.  **Juntamos todo:**
    La demostración completa es:
    $$ \mathcal{L}\left\{\frac{dx(t)}{dt}\right\} = sX(s) - x(0) $$
    Como puedes ver, si la condición inicial es cero, es decir, **si $x(0)=0$**, la fórmula se simplifica a la que te pidieron demostrar:
    $$ \mathcal{L}\left\{\frac{dx(t)}{dt}\right\} = sX(s) $$

---

### iv) Propiedad de Convolución: $\mathcal{L}\{x(t) * y(t)\} = X(s)Y(s)$

La convolución puede parecer intimidante, pero su transformada es muy elegante.

1.  **Recordemos la definición de convolución:**
    La convolución de dos funciones causales $x(t)$ e $y(t)$ es:
    $$ x(t) * y(t) = \int_0^t x(\tau) y(t-\tau) d\tau $$

2.  **Aplicamos la definición de la transformada de Laplace:**
    $$ \mathcal{L}\{x(t) * y(t)\} = \int_0^\infty \left[ \int_0^t x(\tau) y(t-\tau) d\tau \right] e^{-st} dt $$

3.  **Cambiamos el orden de integración:**
    Este es el paso clave. Estamos integrando en una región del plano $t-\tau$. Cambiar el orden de las integrales nos simplificará la vida. La región de integración es $0 \le \tau \le t$ y $0 \le t < \infty$. Si lo reordenamos, integramos primero respecto a $t$ desde $\tau$ hasta $\infty$, y luego respecto a $\tau$ desde $0$ hasta $\infty$.
    $$ \int_0^\infty \int_\tau^\infty \left[ x(\tau) y(t-\tau) d\tau \right] e^{-st} dt = \int_0^\infty x(\tau) \left[ \int_\tau^\infty y(t-\tau) e^{-st} dt \right] d\tau $$

4.  **Resolvemos la integral de adentro:**
    Fijémonos en la integral interna: $\int_\tau^\infty y(t-\tau) e^{-st} dt$.
    Hacemos un cambio de variable. Sea $u = t - \tau$.
    * Entonces $t = u + \tau$ y $dt = du$.
    * Los nuevos límites son: si $t=\tau \implies u=0$, y si $t \to \infty \implies u \to \infty$.
    La integral interna se convierte en:
    $$ \int_0^\infty y(u) e^{-s(u+\tau)} du = \int_0^\infty y(u) e^{-su} e^{-s\tau} du $$
    Sacamos $e^{-s\tau}$ (es constante respecto a $u$):
    $$ e^{-s\tau} \int_0^\infty y(u) e^{-su} du $$
    La integral es la definición de la transformada de $y(t)$, es decir, $Y(s)$. Así que la integral interna es $e^{-s\tau} Y(s)$.

5.  **Sustituimos de vuelta en la integral de afuera:**
    Ahora ponemos este resultado de nuevo en la expresión del paso 3:
    $$ \mathcal{L}\{x(t) * y(t)\} = \int_0^\infty x(\tau) [e^{-s\tau} Y(s)] d\tau $$

6.  **Resultado final:**
    $Y(s)$ es una constante con respecto a $\tau$, así que también puede salir de la integral:
    $$ Y(s) \int_0^\infty x(\tau) e^{-s\tau} d\tau $$
    Y la integral que queda es, por supuesto, la definición de $X(s)$. Por lo tanto, llegamos al resultado final:
    $$ \mathcal{L}\{x(t) * y(t)\} = X(s)Y(s) $$

## Parte 2: Transformada, Polos y Ceros, y ROC de Señales


Para cada señal, encontraremos tres cosas:
1.  La **Transformada de Laplace**, $X(s)$.
2.  El **esquema de polos y ceros** en el plano s.
3.  La **Región de Convergencia (ROC)**.

Recordemos las transformadas básicas que usaremos:
* Para una señal causal (hacia la derecha): $\mathcal{L}\{e^{-at}u(t)\} = \frac{1}{s+a}$, con ROC: $Re(s) > -a$.
* Para una señal anticausal (hacia la izquierda): $\mathcal{L}\{e^{-at}u(-t)\} = -\frac{1}{s+a}$, con ROC: $Re(s) < -a$.

---

### i) $x(t) = e^{-2t}u(t) + e^{-3t}u(t)$

Esta es la suma de dos señales exponenciales que van hacia la derecha.

1.  **Transformada de Laplace:**
    Usamos la propiedad de linealidad. Transformamos cada parte por separado:
    * $\mathcal{L}\{e^{-2t}u(t)\} = \frac{1}{s+2}$, con ROC: $Re(s) > -2$.
    * $\mathcal{L}\{e^{-3t}u(t)\} = \frac{1}{s+3}$, con ROC: $Re(s) > -3$.

    Ahora sumamos las transformadas:
    $X(s) = \frac{1}{s+2} + \frac{1}{s+3}$
    Para tener una sola expresión, combinamos las fracciones:
    $$ X(s) = \frac{(s+3) + (s+2)}{(s+2)(s+3)} = \frac{2s+5}{s^2+5s+6} $$

2.  **Polos y Ceros:**
    * **Polos:** Son las raíces del denominador, los valores de $s$ que hacen la transformada infinita. Se encuentran en $s+2=0$ y $s+3=0$.
        * Polo en $s = -2$.
        * Polo en $s = -3$.
    * **Ceros:** Son las raíces del numerador, los valores de $s$ que hacen la transformada cero. Se encuentran en $2s+5=0$.
        * Cero en $s = -2.5$.

3.  **Región de Convergencia (ROC):**
    La ROC de una suma de señales es la **intersección** de las ROC individuales.
    * ROC de la primera parte: $Re(s) > -2$.
    * ROC de la segunda parte: $Re(s) > -3$.
    La región que cumple ambas condiciones es la más restrictiva. Por lo tanto, la ROC final es $Re(s) > -2$. Esta es una región que se extiende hacia la derecha del polo más a la derecha.

### ii) $x(t) = e^{2t}u(t) + e^{-3t}u(-t)$

Esta señal es "bilateral" o de dos lados: una parte va hacia la derecha ($u(t)$) y la otra hacia la izquierda ($u(-t)$).

1.  **Transformada de Laplace:**
    Transformamos cada parte:
    * Parte derecha: $\mathcal{L}\{e^{2t}u(t)\} = \mathcal{L}\{e^{-(-2)t}u(t)\} = \frac{1}{s-2}$, con ROC: $Re(s) > 2$.
    * Parte izquierda: $\mathcal{L}\{e^{-3t}u(-t)\} = -\frac{1}{s+3}$, con ROC: $Re(s) < -3$.

    La transformada total es la suma:
    $X(s) = \frac{1}{s-2} - \frac{1}{s+3}$
    Combinando las fracciones:
    $$ X(s) = \frac{(s+3) - (s-2)}{(s-2)(s+3)} = \frac{5}{s^2+s-6} $$

2.  **Polos y Ceros:**
    * **Polos:** Son las raíces del denominador: $(s-2)(s+3)=0$.
        * Polo en $s = 2$.
        * Polo en $s = -3$.
    * **Ceros:** El numerador es una constante (5), por lo que no hay ceros finitos.

3.  **Región de Convergencia (ROC):**
    De nuevo, la ROC es la intersección de las ROC individuales:
    * ROC de la parte derecha: $Re(s) > 2$.
    * ROC de la parte izquierda: $Re(s) < -3$.
    Debemos encontrar los valores de $s$ que son al mismo tiempo mayores que 2 y menores que -3. **No existe ningún número que cumpla esto**.

    **Conclusión:** Las regiones de convergencia no se superponen. Por lo tanto, la Transformada de Laplace para esta señal **no existe**.

---

### iii) $x(t) = e^{-a|t|}$ (asumiendo $a > 0$)

Esta señal también es bilateral, gracias al valor absoluto.

1.  **Transformada de Laplace:**
    Primero, reescribimos la señal usando las funciones escalón:
    $x(t) = e^{-at}u(t) + e^{at}u(-t)$
    * Parte derecha: $\mathcal{L}\{e^{-at}u(t)\} = \frac{1}{s+a}$, con ROC: $Re(s) > -a$.
    * Parte izquierda: $\mathcal{L}\{e^{at}u(-t)\} = -\frac{1}{s-a}$, con ROC: $Re(s) < a$.

    Sumando las transformadas:
    $$ X(s) = \frac{1}{s+a} - \frac{1}{s-a} = \frac{(s-a) - (s+a)}{(s+a)(s-a)} = \frac{-2a}{s^2-a^2} $$

2.  **Polos y Ceros:**
    * **Polos:** Son las raíces de $s^2-a^2=0$.
        * Polo en $s = a$.
        * Polo en $s = -a$.
    * **Ceros:** El numerador es una constante ($-2a$), así que no hay ceros finitos.

3.  **Región de Convergencia (ROC):**
    Intersectamos las ROC individuales: $Re(s) > -a$ y $Re(s) < a$.
    La región que cumple ambas es la franja vertical entre $-a$ y $a$.
    ROC: $-a < Re(s) < a$.

### iv) $x(t) = e^{-2t}[u(t) - u(t-5)]$

Esta señal representa una exponencial $e^{-2t}$ que "se enciende" en $t=0$ y "se apaga" en $t=5$. Es una señal de duración finita.

1.  **Transformada de Laplace:**
    La forma más directa es usar la definición de la transformada, ya que los límites de integración son finitos:
    $X(s) = \int_0^\infty x(t) e^{-st} dt = \int_0^5 e^{-2t} e^{-st} dt$
    $X(s) = \int_0^5 e^{-(s+2)t} dt$
    Resolviendo la integral:
    $$ X(s) = \left[ \frac{e^{-(s+2)t}}{-(s+2)} \right]_0^5 = \frac{e^{-5(s+2)}}{- (s+2)} - \frac{e^{0}}{-(s+2)} $$
    $$ X(s) = \frac{1 - e^{-5(s+2)}}{s+2} $$

2.  **Polos y Ceros:**
    * **Polos:** A primera vista, parece que hay un polo en $s=-2$ por el denominador. Sin embargo, veamos si el numerador también es cero en ese punto.
        Si $s=-2$, el numerador es $1 - e^{-5(-2+2)} = 1 - e^0 = 1-1 = 0$.
        Como tanto el numerador como el denominador son cero en $s=-2$, tenemos una "cancelación polo-cero". El valor de la transformada en $s=-2$ es finito (se puede encontrar con la regla de L'Hôpital y da 5). Por lo tanto, **no hay polos**.
    * **Ceros:** Son los valores de $s$ que hacen el numerador cero (excepto $s=-2$).
        $1 - e^{-5(s+2)} = 0 \implies e^{-5(s+2)} = 1$
        Sabemos que $e^{j2\pi k} = 1$ para cualquier entero $k$.
        Entonces: $-5(s+2) = j2\pi k \implies s+2 = -\frac{j2\pi k}{5}$
        $s = -2 - \frac{j2\pi k}{5}$, para $k = \pm 1, \pm 2, \pm 3, ...$
        Hay un número infinito de ceros alineados verticalmente en $Re(s)=-2$.

3.  **Región de Convergencia (ROC):**
    Una propiedad fundamental de la Transformada de Laplace es que si una señal tiene **duración finita**, su integral de Laplace converge para todos los valores de $s$.
    Por lo tanto, la ROC es **todo el plano s**. Esto es consistente con el hecho de que no hay polos que limiten la región.

## Parte 3: Funciones de Transferencia de Circuitos RLC en Lazo Abierto



¿Por qué? Porque la función de transferencia $H(s)$ describe la relación intrínseca entre la salida y la entrada ($H(s) = Y(s)/X(s)$), independientemente del estado en que se encontraba el sistema al principio. Representa la respuesta del sistema "puro" a una entrada. La respuesta total del sistema sí depende de las condiciones iniciales, pero la función de transferencia no.

### Análisis de la Ecuación que Enviaste

La ecuación diferencial que muestras es:
$$CL \frac{d^2y(t)}{dt^2} + CR \frac{dy(t)}{dt} + y(t) = x(t)$$
Vamos a ver a cuál de los dos circuitos corresponde. Para ello, deduciremos la ecuación de cada uno desde cero.

---

### 1. Circuito RLC Serie

**Objetivo:** Encontrar la función de transferencia $H(s) = \frac{V_C(s)}{V_{in}(s)}$, donde la salida es la tensión en el capacitor.

**Análisis del Circuito:**
1.  Imaginemos una fuente de tensión $v_{in}(t)$ conectada en serie a una resistencia $R$, un inductor $L$ y un capacitor $C$.
2.  Aplicamos la **Ley de Voltajes de Kirchhoff (LVK)** alrededor de la malla. La suma de las caídas de tensión en los componentes es igual a la tensión de la fuente:
    $$v_{in}(t) = v_R(t) + v_L(t) + v_C(t)$$
3.  Expresamos cada tensión en términos de la corriente de la malla, $i(t)$:
    * $v_R(t) = R \cdot i(t)$
    * $v_L(t) = L \frac{di(t)}{dt}$
    * $v_C(t)$ es nuestra salida.

4.  La relación clave entre la corriente y el voltaje del capacitor es:
    $$i(t) = C \frac{dv_C(t)}{dt}$$
5.  Ahora, sustituimos $i(t)$ en las expresiones de voltaje para que todo quede en función de nuestra salida $v_C(t)$:
    * $v_R(t) = R \left( C \frac{dv_C(t)}{dt} \right) = RC \frac{dv_C(t)}{dt}$
    * Para $v_L(t)$, primero necesitamos la derivada de la corriente: $\frac{di(t)}{dt} = C \frac{d^2v_C(t)}{dt^2}$.
    * $v_L(t) = L \left( C \frac{d^2v_C(t)}{dt^2} \right) = LC \frac{d^2v_C(t)}{dt^2}$

6.  Sustituimos todo en la ecuación de la LVK:
    $$v_{in}(t) = \left( RC \frac{dv_C(t)}{dt} \right) + \left( LC \frac{d^2v_C(t)}{dt^2} \right) + v_C(t)$$
7.  Reordenando los términos, obtenemos la ecuación diferencial del sistema:
    $$LC \frac{d^2v_C(t)}{dt^2} + RC \frac{dv_C(t)}{dt} + v_C(t) = v_{in}(t)$$
    Si llamamos a la entrada $x(t) = v_{in}(t)$ y a la salida $y(t) = v_C(t)$, la ecuación es idéntica a la que tú proporcionaste. ¡Excelente!

**Cálculo de la Función de Transferencia:**

1.  Aplicamos la Transformada de Laplace a la ecuación diferencial (con condiciones iniciales cero):
    $\mathcal{L}\left\{ LC \frac{d^2y(t)}{dt^2} + RC \frac{dy(t)}{dt} + y(t) \right\} = \mathcal{L}\{x(t)\}$
    $LC \cdot s^2Y(s) + RC \cdot sY(s) + Y(s) = X(s)$
2.  Factorizamos $Y(s)$:
    $Y(s) (LCs^2 + RCs + 1) = X(s)$
3.  Despejamos la relación $Y(s)/X(s)$:
    $$H(s) = \frac{Y(s)}{X(s)} = \frac{V_C(s)}{V_{in}(s)} = \frac{1}{LCs^2 + RCs + 1}$$

Esta es la función de transferencia para el circuito RLC serie con salida en el capacitor.

---

### 2. Circuito RLC Paralelo

**Objetivo:** Encontrar la función de transferencia $H(s) = \frac{I_L(s)}{I_{in}(s)}$, donde la salida es la corriente en el inductor.

**Análisis del Circuito:**
1.  Imaginemos una fuente de corriente $i_{in}(t)$ que alimenta en paralelo a una resistencia $R$, un inductor $L$ y un capacitor $C$.
2.  Aplicamos la **Ley de Corrientes de Kirchhoff (LCK)** en el nodo superior. La corriente que entra es igual a la suma de las que salen:
    $$i_{in}(t) = i_R(t) + i_L(t) + i_C(t)$$
3.  La tensión $v(t)$ es la misma para todos los componentes. Expresamos cada corriente en términos de esta tensión:
    * $i_R(t) = \frac{v(t)}{R}$
    * $i_L(t)$ es nuestra salida.
    * $i_C(t) = C \frac{dv(t)}{dt}$

4.  La relación clave entre el voltaje y la corriente del inductor es:
    $$v(t) = L \frac{di_L(t)}{dt}$$
5.  Sustituimos $v(t)$ en las otras expresiones de corriente para que todo quede en función de nuestra salida $i_L(t)$:
    * $i_R(t) = \frac{1}{R} \left( L \frac{di_L(t)}{dt} \right) = \frac{L}{R} \frac{di_L(t)}{dt}$
    * Para $i_C(t)$, primero necesitamos la derivada del voltaje: $\frac{dv(t)}{dt} = L \frac{d^2i_L(t)}{dt^2}$.
    * $i_C(t) = C \left( L \frac{d^2i_L(t)}{dt^2} \right) = LC \frac{d^2i_L(t)}{dt^2}$

6.  Sustituimos todo en la ecuación de la LCK:
    $$i_{in}(t) = \left( \frac{L}{R} \frac{di_L(t)}{dt} \right) + i_L(t) + \left( LC \frac{d^2i_L(t)}{dt^2} \right)$$
7.  Reordenando, obtenemos la ecuación diferencial (nota que es diferente a la de la imagen):
    $$LC \frac{d^2i_L(t)}{dt^2} + \frac{L}{R} \frac{di_L(t)}{dt} + i_L(t) = i_{in}(t)$$

**Cálculo de la Función de Transferencia:**

1.  Aplicamos la Transformada de Laplace a la ecuación (con condiciones iniciales cero):
    $\mathcal{L}\left\{ LC \frac{d^2y(t)}{dt^2} + \frac{L}{R} \frac{dy(t)}{dt} + y(t) \right\} = \mathcal{L}\{x(t)\}$
    $LC \cdot s^2Y(s) + \frac{L}{R} \cdot sY(s) + Y(s) = X(s)$
2.  Factorizamos $Y(s)$:
    $Y(s) \left(LCs^2 + \frac{L}{R}s + 1\right) = X(s)$
3.  Despejamos la relación $Y(s)/X(s)$:
    $$H(s) = \frac{Y(s)}{X(s)} = \frac{I_L(s)}{I_{in}(s)} = \frac{1}{LCs^2 + \frac{L}{R}s + 1}$$
    Para presentarla en un formato estándar (con $s^2$ solo), podemos dividir todo por $LC$:
    $$H(s) = \frac{\frac{1}{LC}}{s^2 + \frac{1}{RC}s + \frac{1}{LC}}$$

Esta es la función de transferencia para el circuito RLC paralelo con salida en la corriente del inductor.

### Resumen Final

* **Circuito RLC Serie (Salida $V_C$):**
    $$H(s) = \frac{1}{LCs^2 + RCs + 1}$$
    (Esta corresponde a la ecuación diferencial que enviaste).

* **Circuito RLC Paralelo (Salida $I_L$):**
    $$H(s) = \frac{\frac{1}{LC}}{s^2 + \frac{1}{RC}s + \frac{1}{LC}}$$

## Parte 4: Análisis en Lazo Cerrado de Circuitos RLC con Python



### Concepto Clave: Lazo Cerrado con Retroalimentación Unitaria

Para encontrar la función de transferencia en lazo cerrado ($H_{LC}(s)$), asumiremos la configuración más común: **retroalimentación negativa unitaria**. En esta configuración, la salida se resta de la señal de referencia y la diferencia (el error) se alimenta al sistema.

La fórmula para pasar de lazo abierto ($H_{LA}(s)$) a lazo cerrado es:

$$H_{LC}(s) = \frac{H_{LA}(s)}{1 + H_{LA}(s)}$$

El flujo de trabajo será el siguiente para cada circuito:
1.  Calcular la función de transferencia en lazo cerrado, $H_{LC}(s)$.
2.  Para cada entrada $x(t)$ (impulso, escalón, rampa), encontrar la transformada de la salida: $Y(s) = H_{LC}(s) \cdot X(s)$.
3.  Usar `sympy` para encontrar la transformada inversa de Laplace de $Y(s)$ y obtener la expresión de la salida en el tiempo, $y(t)$.
4.  Elegir valores numéricos para R, L y C para poder graficar las respuestas. La forma de la respuesta (subamortiguada, críticamente amortiguada o sobreamortiguada) depende críticamente de estos valores. Elegiremos valores que muestren un comportamiento oscilatorio (subamortiguado) para que las gráficas sean más ilustrativas.

### Preparación del Entorno en Python

Primero, vamos a preparar nuestro entorno importando las librerías necesarias y definiendo los símbolos que usaremos con `sympy`.



```python

import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Configurar sympy para una impresión bonita (en entornos como Jupyter)
sp.init_printing()

# Definir los símbolos que usaremos
s, t = sp.symbols('s t', positive=True)
R, L, C = sp.symbols('R L C', positive=True)

# --- Circuito Serie: Definición de H_LC(s) ---
H_la1 = 1 / (L*C*s**2 + R*C*s + 1)
H_lc1 = H_la1 / (1 + H_la1)
# Simplificamos la expresión
H_lc1 = sp.simplify(H_lc1)

print("Función de Transferencia en Lazo Cerrado (Serie):")
display(H_lc1)

# i) Entrada IMPULSO: X(s) = 1
Y_impulso1 = H_lc1
y_impulso1 = sp.inverse_laplace_transform(Y_impulso1, s, t)
print("\nSalida en el tiempo para entrada IMPULSO y_i(t):")
display(y_impulso1)

# ii) Entrada ESCALÓN: X(s) = 1/s
Y_escalon1 = H_lc1 / s
y_escalon1 = sp.inverse_laplace_transform(Y_escalon1, s, t)
print("\nSalida en el tiempo para entrada ESCALÓN y_e(t):")
display(y_escalon1)

# iii) Entrada RAMPA: X(s) = 1/s^2
Y_rampa1 = H_lc1 / s**2
y_rampa1 = sp.inverse_laplace_transform(Y_rampa1, s, t)
print("\nSalida en el tiempo para entrada RAMPA y_r(t):")
display(y_rampa1)

# Valores numéricos para la simulación
vals = {L: 1, C: 1, R: 1}

# Convertir las expresiones simbólicas de sympy a funciones numéricas de numpy
y_impulso1_num = sp.lambdify(t, y_impulso1.subs(vals), 'numpy')
y_escalon1_num = sp.lambdify(t, y_escalon1.subs(vals), 'numpy')
y_rampa1_num = sp.lambdify(t, y_rampa1.subs(vals), 'numpy')

# Vector de tiempo para la gráfica
t_vec = np.linspace(0, 20, 500)

# Calcular las salidas numéricas
y_i1 = y_impulso1_num(t_vec)
y_e1 = y_escalon1_num(t_vec)
y_r1 = y_rampa1_num(t_vec)

# Graficar los resultados
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(t_vec, y_i1, label='Respuesta al Impulso')
plt.title('Circuito Serie: Entrada Impulso')
plt.xlabel('Tiempo (s)')
plt.ylabel('Salida y(t)')
plt.grid(True)
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(t_vec, y_e1, label='Respuesta al Escalón')
plt.axhline(1/2, color='r', linestyle='--', label='Valor final = 1/2')
plt.title('Circuito Serie: Entrada Escalón')
plt.xlabel('Tiempo (s)')
plt.grid(True)
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(t_vec, y_r1, label='Respuesta a la Rampa')
plt.plot(t_vec, 0.5 * t_vec - 0.25, 'r--', label='Asíntota') # Error en estado estable
plt.title('Circuito Serie: Entrada Rampa')
plt.xlabel('Tiempo (s)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()


# --- Circuito Paralelo: Definición de H_LC(s) ---
H_la2 = (1/(L*C)) / (s**2 + (1/(R*C))*s + (1/(L*C)))
H_lc2 = sp.simplify(H_la2 / (1 + H_la2))

print("Función de Transferencia en Lazo Cerrado (Paralelo):")
display(H_lc2)

# i) Entrada IMPULSO: X(s) = 1
Y_impulso2 = H_lc2
y_impulso2 = sp.inverse_laplace_transform(Y_impulso2, s, t)
print("\nSalida en el tiempo para entrada IMPULSO y_i(t):")
display(y_impulso2)

# ii) Entrada ESCALÓN: X(s) = 1/s
Y_escalon2 = H_lc2 / s
y_escalon2 = sp.inverse_laplace_transform(Y_escalon2, s, t)
print("\nSalida en el tiempo para entrada ESCALÓN y_e(t):")
display(y_escalon2)

# iii) Entrada RAMPA: X(s) = 1/s^2
Y_rampa2 = H_lc2 / s**2
y_rampa2 = sp.inverse_laplace_transform(Y_rampa2, s, t)
print("\nSalida en el tiempo para entrada RAMPA y_r(t):")
display(y_rampa2)

# Valores numéricos para la simulación
vals = {L: 1, C: 1, R: 1}

# Convertir las expresiones simbólicas a funciones numéricas
y_impulso2_num = sp.lambdify(t, y_impulso2.subs(vals), 'numpy')
y_escalon2_num = sp.lambdify(t, y_escalon2.subs(vals), 'numpy')
y_rampa2_num = sp.lambdify(t, y_rampa2.subs(vals), 'numpy')

# Vector de tiempo
t_vec = np.linspace(0, 20, 500)

# Calcular las salidas numéricas
y_i2 = y_impulso2_num(t_vec)
y_e2 = y_escalon2_num(t_vec)
y_r2 = y_rampa2_num(t_vec)

# Graficar los resultados
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(t_vec, y_i2, label='Respuesta al Impulso')
plt.title('Circuito Paralelo: Entrada Impulso')
plt.xlabel('Tiempo (s)')
plt.ylabel('Salida y(t)')
plt.grid(True)
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(t_vec, y_e2, label='Respuesta al Escalón')
plt.axhline(1/2, color='r', linestyle='--', label='Valor final = 1/2')
plt.title('Circuito Paralelo: Entrada Escalón')
plt.xlabel('Tiempo (s)')
plt.grid(True)
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(t_vec, y_r2, label='Respuesta a la Rampa')
plt.plot(t_vec, 0.5 * t_vec - 0.25, 'r--', label='Asíntota')
plt.title('Circuito Paralelo: Entrada Rampa')
plt.xlabel('Tiempo (s)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
